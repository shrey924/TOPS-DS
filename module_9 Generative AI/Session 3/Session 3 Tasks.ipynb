```python
# 1
import numpy as np

def simple_generator(random_vector):
    # Convert random vector into 784 values
    pixels = np.tanh(random_vector[:784])

    # Scale values from [-1, 1] to [0, 1]
    pixels = (pixels + 1) / 2

    # Reshape into a 28x28 grayscale image
    image = pixels.reshape(28, 28)

    return image

# Example usage
random_vector = np.random.randn(784)
fake_digit = simple_generator(random_vector)

print(fake_digit.shape)  # (28, 28)
print(fake_digit.min(), fake_digit.max())  # Values between 0 and 1
```

```python
# 2 
import numpy as np

class Discriminator:
    def discriminate(self, image):
        # Calculate mean pixel value
        mean_value = np.mean(image)

        # Return a probability score between 0 and 1
        probability = min(max(mean_value, 0), 1)

        return probability

# Example usage
image = np.random.rand(28, 28)  # Fake 28x28 grayscale image

d = Discriminator()
score = d.discriminate(image)

print("Probability of being real:", score)
```

```python
# 3
import numpy as np

def simple_generator():
    return np.random.rand(28, 28)

class Discriminator:
    def discriminate(self, image):
        return np.mean(image)

# One training cycle
fake_image = simple_generator()
score = Discriminator().discriminate(fake_image)

print("Score:", score)
print("Correctly identified as FAKE" if score < 0.5 else "Classified as REAL")
```

1.DCGAN (Deep Convolutional GAN) uses convolutional neural networks to generate images and is one of the earliest GAN architectures for stable image generation.
2.StyleGAN improves image quality by controlling different visual styles at multiple levels, producing highly realistic images with fine details.
