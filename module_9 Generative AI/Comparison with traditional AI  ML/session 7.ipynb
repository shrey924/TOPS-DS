2. Given a scenario where Zomato recommends restaurants based on your previous orders, explain whether this uses traditional AI/ML or Gen AI, and justify your answer in 3-4 sentences.

Zomato recommending restaurants based on previous orders is mainly an example of Traditional AI/ML. The system analyzes a user's past behavior, preferences, ratings, and ordering history to predict which restaurants or dishes the user might like. It does not create new content; instead, it selects and ranks existing restaurants. Therefore, it is a recommendation system based on machine learning rather than generative AI.

3. Pick any feature from an app you use daily (such as Instagram's photo filters or Spotify's song recommendations) and describe how it could be improved or transformed using Gen AI instead of traditional AI/ML. Write 4-5 lines explaining the change.

Feature: Spotify Song Recommendations

Currently, Spotify recommends songs based on listening history and user preferences. With Gen AI, Spotify could generate a personalized playlist description, explain why each song was selected, and even create custom transitions between songs. It could also generate unique playlists for specific moods or events using natural language prompts such as "Create a playlist for a rainy evening study session." This would make the experience more interactive and personalized than traditional recommendation systems.

4. Imagine you are building a chatbot for Flipkart customer support. List two ways a traditional AI/ML approach would handle user queries, and two ways a Gen AI approach would differ in handling the same queries.<br><br><em><strong>

Traditional AI/ML Approach
Uses predefined FAQs and rule-based responses for common questions.
Classifies user queries into categories (refund, order status, return request) and provides fixed answers.
Gen AI Approach
Generates natural, conversational responses tailored to the user's specific question and context.
Can understand complex or multi-part queries and provide detailed explanations without relying solely on predefined templates.
Example

User: "My order is delayed and I want to know if I can get a refund."

Traditional AI/ML: Returns a fixed response from the refund policy database.
Gen AI: Understands both the delay and refund concerns, explains the current order status, summarizes the refund policy, and suggests the next steps in a conversational manner.
