```python
class BankAccount:
    def __init__(self,balance: float = 0.0) ->None:
        self.balance = balance

    def deposite(self, amount: float) ->None:
        if amount > 0:
            self.balance += amount
            print(f"Deposite ${amount:.2f}. current balance: ${self.balance:.2f}")

        else:
            print("deposit amount must be positive.")

    def withdraw(self, amount: float) ->None:
        if amount > 0:
            if self.balance >= amount:
                self.balance -= amount
                print(f"withdraw ${amount:.2f}. current balance: ${self.balance:.2f}")
            else:
                print(f"insufficient fund. current balance: ${self.balance:.2f}")        
        else:
            print("withdraw amount must be positive.")

    def Check_Blance (self) ->None:
        print(f"current balance: ${self.balance:.2f}")


def main() -> None:

    account = BankAccount(1000.0)

    while True:
        print("\n -- Banking system--")
        print("1. Deposit")
        print("2. withdraw")
        print("3. Check Blance")
        print("4. Exit")

        choice = input("Choose an option: ")

        if choice == '1':
            try:
                amount = float(input("enter amount to deposit: $"))
                account.deposite(amount)
            except ValueError:
                print("Invalid input ! enter a valid number of the deposit.")
        elif choice == '2':
            try:
                amount = float(input("enter amount to withraw: $"))
                account.withdraw(amount)
            except ValueError:
                print("Invalid input ! enter a valid number of the withdrawal.")
        elif choice == '3':
            account.Check_Blance()
        elif choice == '4':
            print("Exiting the banking system. Goodbye ")
            break

        else:
            print("Invalid optin please try again. ")

if __name__ == "__main__":
    main()

```
