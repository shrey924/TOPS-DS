```python
# 1 
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    
    def get_discounted_price(self):
        return self.price * 0.90  
product = Product("Smartphone", 20000)

print("Product Name:", product.name)
print("Original Price: ₹", product.price)
print("Discounted Price: ₹", product.get_discounted_price())
```

```python
# 3

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

   
    def get_discounted_price(self):
        return self.price * 0.90



class Electronics(Product):

    
    def get_discounted_price(self):
        return self.price * 0.80

def show_final_price(item):
    print(f"Product Name: {item.name}")
    print(f"Discounted Price: ₹{item.get_discounted_price()}")
    print()

product = Product("Notebook", 100)
electronic = Electronics("Laptop", 50000)

show_final_price(product)
show_final_price(electronic)
```

```python
# 4

class Ticket:
    def __init__(self, movie_name, price):
        self.movie_name = movie_name
        self.price = price

   
    def get_final_price(self):
        return self.price



class PremiumTicket(Ticket):

   
    def get_final_price(self):
        return super().get_final_price() + 50  



regular_ticket = Ticket("Avengers: Endgame", 250)
premium_ticket = PremiumTicket("Avengers: Endgame", 250)
print("Movie:", regular_ticket.movie_name)
print("Regular Ticket Price: ₹", regular_ticket.get_final_price())
print()
print("Movie:", premium_ticket.movie_name)
print("Premium Ticket Price: ₹", premium_ticket.get_final_price())
```
