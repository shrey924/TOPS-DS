```python
# 1 
import requests


url = "https://jsonplaceholder.typicode.com/posts"


payload = {
    "title": "Learning Python Requests",
    "body": "This is my first POST request using the requests library.",
    "userId": 1
}


response = requests.post(url, json=payload)
print("Status Code:", response.status_code)
print("Response JSON:")
print(response.json())
```

```python
# 2
import requests


url = "https://jsonplaceholder.typicode.com/posts"
playlist_name = input("Enter playlist name: ")
playlist_description = input("Enter playlist description: ")

payload = {
    "title": playlist_name,
    "body": playlist_description,
    "userId": 1
}

response = requests.post(url, json=payload)

if response.status_code == 201:
    data = response.json()
    print("\nPlaylist created successfully!")
    print("Playlist ID:", data["id"])
else:
    print("Failed to create playlist.")
    print("Status Code:", response.status_code)
```

```python
# 3
import requests

# API endpoint
url = "https://reqres.in/api/users"
payload = {
    "username": "john_doe",
    "job": "Software Developer"
}


headers = {
    "x-api-key": "reqres-free-v1"
}

response = requests.post(url, json=payload, headers=headers)

if response.status_code == 201:
    data = response.json()

    print("User created successfully!")
    print("User ID:", data["id"])
    print("Created At:", data["createdAt"])
else:
    print("Request failed!")
    print("Status Code:", response.status_code)
    print("Response:", response.text)
```

```python
# 4
import requests
import csv


url = "https://jsonplaceholder.typicode.com/posts"

response = requests.get(url)


if response.status_code == 200:
    posts = response.json()

   
    with open("posts.csv", "w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow(["Title", "User ID"])
        for post in posts[:5]:
            writer.writerow([post["title"], post["userId"]])

    print("Data has been saved to posts.csv")
else:
    print("Request failed with status code:", response.status_code)
```

```python
# 5
import requests
import json


url = "https://jsonplaceholder.typicode.com/posts"
response = requests.get(url)


if response.status_code == 200:
    posts = response.json()
    latest_posts = posts[:5]

   
    with open("posts.json", "w", encoding="utf-8") as file:
        json.dump(latest_posts, file, indent=4)

    print("Data has been saved to posts.json")
else:
    print("Request failed with status code:", response.status_code)
```
