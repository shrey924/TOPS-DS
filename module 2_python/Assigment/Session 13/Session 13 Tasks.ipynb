```python
# 1 

def print_playlist_songs(songs):
    if not songs:
        return
    print(songs[0])
    print_playlist_songs(songs[1:])
playlist = ["Perfect","Believer","Shape of You","Blinding Lights","Levitating"]
print("Spotify Playlist:")
print_playlist_songs(playlist)
```

```python
# 2 

def count_unread_messages(messages):
    total = messages.get("count", 0)
    for subgroup in messages.get("subgroups", []):
        total += count_unread_messages(subgroup)

    return total
whatsapp_groups = {
    "name": "Family",
    "count": 10,
    "subgroups": [
        {
            "name": "Cousins",
            "count": 5,
            "subgroups": []
        },
        {
            "name": "Friends",
            "count": 8,
            "subgroups": [
                {
                    "name": "College Friends",
                    "count": 3,
                    "subgroups": []
                },
                {
                    "name": "Office Friends",
                    "count": 7,
                    "subgroups": []
                }
            ]
        }
    ]
}
total_unread = count_unread_messages(whatsapp_groups)
print("Total Unread Messages:", total_unread)
```

```python
# 3
x = "global"

def outer():
    x = "outer"

    def inner():
        nonlocal x
        x = "inner"

    inner()
    print("Inside outer:", x)

outer()
print("Outside:", x)
```

```python
# 4

def format_number_short(n, suffixes=None):
    if suffixes is None:
        suffixes = ["", "K", "M", "B", "T"]
    if n < 1000 or len(suffixes) == 1:
        if suffixes[0] == "":
            return str(int(n))
        else:
            return f"{n:.1f}{suffixes[0]}"
    return format_number_short(n / 1000, suffixes[1:])


# Example usage
print(format_number_short(500))        # 500
print(format_number_short(1500))       # 1.5K
print(format_number_short(1200000))    # 1.2M
print(format_number_short(2500000000)) # 2.5B
```
