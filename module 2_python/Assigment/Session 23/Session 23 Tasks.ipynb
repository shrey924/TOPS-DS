```python
# 1
import requests


url = "https://api.openweathermap.org/data/2.5/weather"


params = {
    "q": "Ahmedabad",
    "appid": "YOUR_API_KEY",   
    "units": "metric"          
}

# Send GET request
response = requests.get(url, params=params)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()

    city = data["name"]
    temperature = data["main"]["temp"]
    description = data["weather"][0]["description"]

    print(f"Weather Report for {city}")
    print(f"Temperature: {temperature}°C")
    print(f"Condition: {description.capitalize()}")
else:
    print("Failed to fetch weather data.")
    print("Status Code:", response.status_code)
    print("Response:", response.text)
```

```python
# 2
import requests
from datetime import datetime


url = "https://api.coingecko.com/api/v3/simple/price"


params = {
    "ids": "bitcoin,ethereum",
    "vs_currencies": "usd"
}


response = requests.get(url, params=params)


if response.status_code == 200:
    data = response.json()

    bitcoin_price = data["bitcoin"]["usd"]
    ethereum_price = data["ethereum"]["usd"]

    current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    print("Crypto Price Report")
    print("-" * 30)
    print(f"Date & Time : {current_time}")
    print(f"Bitcoin     : ${bitcoin_price:,}")
    print(f"Ethereum    : ${ethereum_price:,}")
else:
    print("Failed to fetch cryptocurrency prices.")
    print("Status Code:", response.status_code)
```

```python
# 3
import requests

api_url = "https://api.nasa.gov/planetary/apod"


params = {
    "api_key": "DEMO_KEY"   
}


response = requests.get(api_url, params=params)

if response.status_code == 200:
    data = response.json()

    
    print("Title:", data["title"])
    print("\nExplanation:")
    print(data["explanation"])

    
    if data["media_type"] == "image":
        image_url = data["url"]

        image_response = requests.get(image_url)

        if image_response.status_code == 200:
            with open("apod_image.jpg", "wb") as file:
                file.write(image_response.content)

            print("\nImage downloaded successfully as 'apod_image.jpg'")
        else:
            print("Failed to download the image.")
    else:
        print("\nToday's APOD is not an image (it may be a video).")

else:
    print("Failed to fetch APOD data.")
    print("Status Code:", response.status_code)
```

```python
# 4
import requests

# API endpoint
url = "https://disease.sh/v3/covid-19/countries/India"

# Send GET request
response = requests.get(url)

# Check if request was successful
if response.status_code == 200:
    data = response.json()

    # Display data in table format
    print("+----------------+----------------+")
    print("| Metric         | Value          |")
    print("+----------------+----------------+")
    print(f"| Total Cases    | {data['cases']:<14}|")
    print(f"| Recovered      | {data['recovered']:<14}|")
    print(f"| Deaths         | {data['deaths']:<14}|")
    print("+----------------+----------------+")
else:
    print("Failed to fetch COVID-19 data.")
    print("Status Code:", response.status_code)
```

```python
# 5
import requests



weather_url = "https://api.openweathermap.org/data/2.5/weather"
weather_params = {
    "q": "Mumbai",
    "appid": "YOUR_API_KEY",   
    "units": "metric"
}


crypto_url = "https://api.coingecko.com/api/v3/simple/price"
crypto_params = {
    "ids": "bitcoin",
    "vs_currencies": "usd"
}

weather_result = None
bitcoin_price = None

try:
    weather_response = requests.get(weather_url, params=weather_params, timeout=10)
    weather_response.raise_for_status()

    weather_data = weather_response.json()
    weather_result = {
        "temperature": weather_data["main"]["temp"],
        "description": weather_data["weather"][0]["description"]
    }

except requests.exceptions.RequestException as e:
    print("Error fetching weather data:", e)


try:
    crypto_response = requests.get(crypto_url, params=crypto_params, timeout=10)
    crypto_response.raise_for_status()

    crypto_data = crypto_response.json()
    bitcoin_price = crypto_data["bitcoin"]["usd"]

except requests.exceptions.RequestException as e:
    print("Error fetching Bitcoin price:", e)


print("\n===== Live Information =====")

if weather_result:
    print(f"Mumbai Temperature : {weather_result['temperature']}°C")
    print(f"Weather            : {weather_result['description'].title()}")
else:
    print("Weather information is unavailable.")

if bitcoin_price:
    print(f"Bitcoin Price      : ${bitcoin_price:,}")
else:
    print("Bitcoin price is unavailable.")
```
