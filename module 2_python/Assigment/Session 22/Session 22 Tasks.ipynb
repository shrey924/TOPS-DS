```python
# 1
import requests

# Flipkart order history URL
url = "https://www.flipkart.com/account/orders"

# Create a session
session = requests.Session()

# First request
response1 = session.get(url)

# Second request
response2 = session.get(url)

# Print status codes
print("First Request Status Code:", response1.status_code)
print("Second Request Status Code:", response2.status_code)

# Print cookies stored in the session
print("\nSession Cookies:")
print(session.cookies.get_dict())

# Print the final URLs (to check if redirected)
print("\nFirst Request URL:", response1.url)
print("Second Request URL:", response2.url)
```

```python
# 2
import requests
url = "https://api.openweathermap.org/data/2.5/weather"


params = {
    "q": "Ahmedabad",
    "appid": "YOUR_API_KEY",   
    "units": "metric"         
}
response = requests.get(url, params=params)


if response.status_code == 200:
    data = response.json()

    temperature = data["main"]["temp"]
    print(f"Current temperature in Ahmedabad: {temperature}°C")
else:
    print("Request failed!")
    print("Status Code:", response.status_code)
    print("Response:", response.text)
```

```python
# 3
import asyncio
import httpx

# Fetch posts (simulating trending songs)
async def fetch_trending_songs():
    url = "https://jsonplaceholder.typicode.com/posts?_limit=5"

    async with httpx.AsyncClient() as client:
        response = await client.get(url)
        response.raise_for_status()
        return response.json()

# Fetch albums (simulating trending movies)
async def fetch_trending_movies():
    url = "https://jsonplaceholder.typicode.com/albums?_limit=5"

    async with httpx.AsyncClient() as client:
        response = await client.get(url)
        response.raise_for_status()
        return response.json()

async def main():
    # Run both requests concurrently
    songs, movies = await asyncio.gather(
        fetch_trending_songs(),
        fetch_trending_movies()
    )

    print("Trending Songs:")
    for song in songs:
        print("-", song["title"])

    print("\nTrending Movies:")
    for movie in movies:
        print("-", movie["title"])

# Run the async program
asyncio.run(main())
```

```python
# 4
import requests

def get_user_profile():
    # API endpoint
    url = "https://jsonplaceholder.typicode.com/users/1"

    # Fake Bearer token
    token = "fake_bearer_token_12345"

    # Request headers
    headers = {
        "Authorization": f"Bearer {token}"
    }

    # Send GET request
    response = requests.get(url, headers=headers)

    # Check if the request was successful
    if response.status_code == 200:
        user = response.json()
        print("User Name:", user["name"])
    else:
        print("Request failed!")
        print("Status Code:", response.status_code)

# Call the function
get_user_profile()
```
