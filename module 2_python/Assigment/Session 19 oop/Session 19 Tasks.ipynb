```python
# 2 

class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration  

    
    def display_details(self):
        print("Title:", self.title)
        print("Artist:", self.artist)
        print("Duration:", self.duration, "seconds")

    
    def play_preview(self):
        print(f"Playing 30-second preview of {self.title} by {self.artist}")


favorite_song = Song("Perfect", "Ed Sheeran", 263)
favorite_song.display_details()
favorite_song.play_preview()
```

```python
# 3 

class FoodOrder:
    def __init__(self, restaurant_name, items, total_price):
        self.restaurant_name = restaurant_name
        self.items = items
        self.total_price = total_price

    
    def display_order(self):
        print("Food Order Details")
        print("Restaurant Name:", self.restaurant_name)
        print("Items:", ", ".join(self.items))
        print("Total Price: ₹", self.total_price)

my_order = FoodOrder(
    "Domino's Pizza",
    ["Margherita Pizza", "Garlic Bread", "Coke"],
    699
)


my_order.display_order()
```

```python
# 4
class FoodOrder:
    def __init__(self, restaurant_name, items, total_price):
        self.restaurant_name = restaurant_name
        self.items = items
        self.total_price = total_price

    
    def add_item(self, item_name, item_price):
        self.items.append(item_name)
        self.total_price += item_price

    
    def display_order(self):
        print("Food Order Details")
        print("Restaurant Name:", self.restaurant_name)
        print("Items:", ", ".join(self.items))
        print("Total Price: ₹", self.total_price)


my_order = FoodOrder(
    "Domino's Pizza",
    ["Margherita Pizza"],
    299
)

my_order.add_item("Garlic Bread", 149)
my_order.add_item("Coke", 50)
my_order.display_order()
```
