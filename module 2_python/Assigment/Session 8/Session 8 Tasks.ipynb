```python
# 2

daily_steps = [6500, 8200, 9800, 10500, 12000, 9500, 11000]
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
i = 0
while i < len(daily_steps):
    if daily_steps[i] > 10000:
        print("First day crossing 10,000 steps:", days[i])
        print("Steps:", daily_steps[i])
        break
    i += 1
```

```python
# 3

def print_long_team_names(team_names):
    for team in team_names:
        if len(team) <= 6:
            continue  
        print(team)

ipl_teams = [
    "CSK",
    "Mumbai Indians",
    "RCB",
    "Sunrisers Hyderabad",
    "Gujarat Titans",
    "KKR",
    "Punjab Kings"
]
print_long_team_names(ipl_teams)
```

```python
# 5

item_prices = [450, 799, 0, 650, 300, 200, 500]
total = 0


for price in item_prices:
    
    if price == 0:
        continue
    total += price
    if total > 2000:
        break

print("Final Cart Total: ₹", total)
```
