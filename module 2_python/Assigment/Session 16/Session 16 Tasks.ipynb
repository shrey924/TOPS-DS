```python
# 2
import random

def playlist_generator(songs):
    shuffled_songs = songs.copy()
    random.shuffle(shuffled_songs)
    for song in shuffled_songs:
        yield song

playlist = ["Perfect","Believer","Shape of You","Blinding Lights","Levitating"]
shuffle_playlist = playlist_generator(playlist)
print("Spotify Shuffle Playlist:")
for song in shuffle_playlist:
    print("Now Playing:", song)
```

```python
# 5

def order_id_generator():
    order_id = 1001
    while True:
        yield order_id
        order_id += 1
orders = order_id_generator()
print("Generated Order IDs:")
print(next(orders))
print(next(orders))
print(next(orders))
print(next(orders))
print(next(orders))
```
