```python
# 3
import csv


with open("ipl_scores.csv", "r") as file:
    reader = csv.DictReader(file)
    print("IPL Match Winners:")
    for row in reader:
        print(f"Match {row['match_id']}: {row['winner']}")
```

```python
# 4
import json
with open("movies.json", "r") as file:
    movies = json.load(file)
print("Trending Movies:")
for movie in movies:
    print(f"Title: {movie['title']}, Rating: {movie['rating']}")
```

```python
# 5
from pathlib import Path
import json

file_path = Path("my_fav_apps.json")


if not file_path.exists():
   
    apps = [
        {"name": "Instagram", "category": "Social Media"},
        {"name": "Zomato", "category": "Food Delivery"},
        {"name": "Paytm", "category": "Digital Payments"}
    ]

   
    with open(file_path, "w") as file:
        json.dump(apps, file, indent=4)

    print("File created successfully!")
else:
    print("File already exists.")
```
