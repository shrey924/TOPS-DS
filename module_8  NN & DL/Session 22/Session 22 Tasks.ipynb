```python
# 1 Load any 3 sample images (your choice: selfies, food, cricket, or memes) and use OpenCV or PIL to resize each to 224x224 pixels, saving the resized versions in a new folder.


from PIL import Image
import os

# Create output folder
os.makedirs("resized_images", exist_ok=True)
images = ["Virat-Kohli-14.jpg", "virat-kohli-ct-3-201445.webp", "virat-kohli-test-696x392.avif"]
for img_file in images:
    img = Image.open(img_file)
    img = img.resize((224, 224))

    save_path = os.path.join("resized_images", img_file)
    img.save(save_path)

    print(f"Saved: {save_path}")
```

```python
# 2 Convert a colored image (choose any photo you like) to grayscale and then back to RGB using Python libraries. Display all three versions (original, grayscale, RGB) using matplotlib.

from PIL import Image
import matplotlib.pyplot as plt

# Load image
img = Image.open("Virat-Kohli-14.jpg")

# Convert to grayscale
gray = img.convert("L")
rgb = gray.convert("RGB")
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title("Original")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(rgb)
plt.title("RGB")
plt.axis("off")

plt.show()
```

```python
# 3 Use Keras' ImageDataGenerator to apply random rotation and horizontal flip to an image of your choice (for example, a Zomato food pic or an IPL player). Generate and display 4 augmented versions.<br><br><em><strong>Hint:</strong> Set rotation_range and horizontal_flip in ImageDataGenerator, then use flow() or flow_from_directory().</em>

from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
import matplotlib.pyplot as plt
import numpy as np

img = np.expand_dims(img_to_array(load_img("virat-kohli-ct-3-201445.webp")), 0)

gen = ImageDataGenerator(rotation_range=30, horizontal_flip=True)

for i, batch in enumerate(gen.flow(img, batch_size=1)):
    plt.subplot(2, 2, i+1)
    plt.imshow(batch[0].astype("uint8"))
    plt.axis("off")
    if i == 3:
        break

plt.show()
```

```python
# 4 Write code to increase and decrease the brightness of a selfie or any image by 40% using ImageDataGenerator's brightness_range parameter. Show the original, brightened, and darkened images side by side.

img = img_to_array(load_img("virat-kohli-test-696x392.avif"))
img = np.expand_dims(img, 0)

# Darken (60% brightness) and Brighten (140% brightness)
dark_gen = ImageDataGenerator(brightness_range=[0.6, 0.6])
bright_gen = ImageDataGenerator(brightness_range=[1.4, 1.4])

dark = next(dark_gen.flow(img, batch_size=1))[0].astype("uint8")
bright = next(bright_gen.flow(img, batch_size=1))[0].astype("uint8")
```

```python
plt.figure(figsize=(9,3))

plt.subplot(1,3,1)
plt.imshow(img[0].astype("uint8"))
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(bright)
plt.title("Bright +40%")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(dark)
plt.title("Dark -40%")
plt.axis("off")

plt.show()
```

```python
# 5 Pick any meme or cricket image and use ImageDataGenerator to apply a shear transformation. Explain in one line how the image changes visually after shearing.

img = np.expand_dims(img_to_array(load_img("image.avif")), 0)

# Shear transformation
gen = ImageDataGenerator(shear_range=30)

sheared = next(gen.flow(img, batch_size=1))[0].astype("uint8")
```
