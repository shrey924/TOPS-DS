```python
# 1 Prepare a sequence dataset using closing prices of any stock (download from Yahoo Finance or use a small CSV sample), and preprocess it into input sequences and targets suitable for training a GRU model in Keras.<br><br><em><strong>Hint:</strong> Use sliding windows to create input sequences of length 10 and the next value as the target.</em>


import numpy as np

prices = [100, 102, 101, 105, 108, 110, 109, 112, 115, 117, 120, 122, 121, 125, 128]

X, y = [], []
window_size = 10

for i in range(len(prices) - window_size):
    X.append(prices[i:i + window_size])
    y.append(prices[i + window_size])

X = np.array(X)
y = np.array(y)
X = X.reshape((X.shape[0], X.shape[1], 1))

print("X shape:", X.shape)
print("y shape:", y.shape)
print("\nFirst Input Sequence:")
print(X[0].flatten())
print("Target:")
print(y[0])
```

```python
# 2 Build and train a simple GRU-based model (using Keras or PyTorch) to predict the next value in your prepared stock price sequence dataset. Print the training loss after each epoch.

import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense


prices = [100, 102, 101, 105, 108, 110, 109, 112, 115, 117,120, 122, 121, 125, 128]
X, y = [], []
window_size = 10

for i in range(len(prices) - window_size):
    X.append(prices[i:i + window_size])
    y.append(prices[i + window_size])

X = np.array(X).reshape(-1, window_size, 1)
y = np.array(y)

model = Sequential([
    GRU(16, input_shape=(window_size, 1)),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
history = model.fit(X, y, epochs=10, verbose=1)
print("\nTraining Loss per Epoch:")
for epoch, loss in enumerate(history.history['loss'], start=1):
    print(f"Epoch {epoch}: Loss = {loss:.4f}")
```

```python
# 3 Now build and train an LSTM-based model (same architecture as your GRU model except for the layer type) on the same dataset. Record the total training time and final accuracy for both GRU and LSTM models.<br><br><em><strong>Hint:</strong> Use Python's time module to measure training duration.</em>

from tensorflow.keras.layers import  LSTM
import time

prices = [100,102,101,105,108,110,109,112,115,117,
          120,122,121,125,128]


X, y = [], []
window = 10

for i in range(len(prices) - window):
    X.append(prices[i:i+window])
    y.append(prices[i+window])

X = np.array(X).reshape(-1, window, 1)
y = np.array(y)
```

```python
gru_model = Sequential([
    GRU(16, input_shape=(window, 1)),
    Dense(1)
])

gru_model.compile(optimizer='adam',
                  loss='mse',
                  metrics=['mae'])

start = time.time()
gru_history = gru_model.fit(X, y, epochs=10, verbose=0)
gru_time = time.time() - start
```

```python
lstm_model = Sequential([
    LSTM(16, input_shape=(window, 1)),
    Dense(1)
])

lstm_model.compile(optimizer='adam',
                   loss='mse',
                   metrics=['mae'])

start = time.time()
lstm_history = lstm_model.fit(X, y, epochs=10, verbose=0)
lstm_time = time.time() - start
```

```python
print("GRU Training Time :", round(gru_time, 2), "seconds")
print("GRU Final MAE     :", round(gru_history.history['mae'][-1], 4))
print("\nLSTM Training Time:", round(lstm_time, 2), "seconds")
print("LSTM Final MAE    :", round(lstm_history.history['mae'][-1], 4))
```

A GRU (Gated Recurrent Unit) is often preferred when:

1.Training speed is important.
2.The dataset is relatively small.
3.The sequence dependencies are not extremely long.
4.You want a simpler model with fewer parameters and lower memory usage.

| Application                                                                            | Why Choose 
| Music Recommendation App (predict next song based on listening history)         | User listening patterns are usually short-term. GRU trains faster and can capture recent preferences effectively.                    |
| Mobile Keyboard Autocomplete** (next-word prediction)                             | Autocomplete needs low latency and fast inference on mobile devices. GRU's smaller size makes it more efficient than LSTM.           |
| E-commerce Product Recommendation** (predict next product a user may view or buy) | User browsing sessions are often short sequences. GRU can model these patterns accurately while using fewer computational resources. |
