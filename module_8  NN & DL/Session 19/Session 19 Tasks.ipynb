```python
# 1 Create a 5x5 NumPy matrix with random integers from 0 to 9, then define a 3x3 filter (kernel) and manually perform a single convolution operation (no padding, stride 1) to produce the activation map.

import numpy as np

img = np.random.randint(0, 10, (5, 5))

kernel = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])

out = np.zeros((3, 3))

for i in range(3):
    for j in range(3):
        out[i, j] = np.sum(img[i:i+3, j:j+3] * kernel)

print("Image:\n", img)
print("\nActivation Map:\n", out)
```

```python
# 2 Implement a Python function apply_stride_convolution(input_matrix, kernel, stride) that performs convolution with a given stride on a 6x6 matrix and a 3x3 kernel, returning the resulting activation map.<br><br><em><strong>Hint:</strong> Use nested loops to move the kernel by the stride value over the input matrix.</em>

from numpy.lib.stride_tricks import sliding_window_view

def apply_stride_convolution(img, kernel, stride):
    windows = sliding_window_view(img, (3, 3))[::stride, ::stride]
    return np.sum(windows * kernel, axis=(2, 3))

# Example
img = np.random.randint(0, 10, (6, 6))
kernel = np.ones((3, 3))

print(apply_stride_convolution(img, kernel, 2))
```

```python
# 3 Modify your convolution code to add zero-padding of 1 around the input matrix before applying the convolution, and show the new activation map output.

import numpy as np
img = np.random.randint(0, 10, (5, 5))

# 3x3 kernel
kernel = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])

# Add zero-padding of 1
padded = np.pad(img, pad_width=1, mode='constant')

# Convolution
out = np.zeros((5, 5))

for i in range(5):
    for j in range(5):
        out[i, j] = np.sum(padded[i:i+3, j:j+3] * kernel)

print("Input:\n", img)
print("\nPadded Input:\n", padded)
print("\nActivation Map:\n", out)
```

```python
# 4 Visualize the effect of different filters by applying both an edge detection kernel and a blur kernel to the same 5x5 matrix and print both activation maps side by side.<br><br><em><strong>Hint:</strong> Use the Sobel or Laplacian kernel for edge detection and an averaging kernel for blur.</em>

import numpy as np

img = np.array([
    [1, 2, 3, 4, 5],
    [5, 4, 3, 2, 1],
    [1, 2, 3, 4, 5],
    [5, 4, 3, 2, 1],
    [1, 2, 3, 4, 5]
])

edge = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

blur = np.ones((3,3)) / 9

edge_map = np.zeros((3,3))
blur_map = np.zeros((3,3))

for i in range(3):
    for j in range(3):
        block = img[i:i+3, j:j+3]
        edge_map[i,j] = np.sum(block * edge)
        blur_map[i,j] = np.sum(block * blur)

print("Edge Detection:\n", edge_map)
print("\nBlur:\n", blur_map)
```
