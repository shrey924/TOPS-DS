```python
# 3 Compile your model using the 'adam' optimizer, 'binary_crossentropy' loss, and include 'accuracy' as a metric.

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(16, activation='relu', input_shape=(2,)),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
# 4 Simulate a small dataset: generate 100 random user profiles (features: listening_time, liked_genre_score) and a binary label (liked_song: 0 or 1). Train your model using model.fit() for 20 epochs and print the accuracy after training.<br><br><em><strong>Hint:</strong> Use numpy to generate random data for features and labels.</em>

import numpy as np

X = np.random.rand(100, 2)  # listening_time, liked_genre_score
y = np.random.randint(0, 2, 100)
```

```python
model = Sequential([
    Dense(16, activation='relu', input_shape=(2,)),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(X, y, epochs=20, verbose=0)
```

```python
print("Accuracy:", history.history['accuracy'][-1])
```

```python
# 5 Add a callback to your fit() call that stops training early if the validation loss does not improve for 3 epochs.<br><br><em><strong>Hint:</strong> Use Keras' EarlyStopping callback and set patience=3.</em>

from tensorflow.keras.callbacks import EarlyStopping

X = np.random.rand(100, 2)
y = np.random.randint(0, 2, 100)
```

```python
model = Sequential([
    Dense(16, activation='relu', input_shape=(2,)),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Early Stopping
early_stop = EarlyStopping(monitor='val_loss',patience=3,restore_best_weights=True)
```

```python
history = model.fit(X, y,epochs=20,validation_split=0.2,callbacks=[early_stop])
```
