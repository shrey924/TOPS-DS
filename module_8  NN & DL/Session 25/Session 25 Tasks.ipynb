```python
# 1 Create a simple Python script using Keras to build an LSTM layer for predicting the next value in a sequence of daily step counts (e.g., [4000, 4200, 4300, 4100, 4500, ...]) like those tracked by a fitness app.

import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense


data = np.array([4000, 4200, 4300, 4100, 4500, 4700, 4600, 4800])

X, y = [], []
window = 3

for i in range(len(data) - window):
    X.append(data[i:i+window])
    y.append(data[i+window])

X = np.array(X)
y = np.array(y)
X = X.reshape((X.shape[0], X.shape[1], 1))

model = Sequential([LSTM(32, input_shape=(window, 1)),Dense(1)])
model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=50, verbose=0)
test_input = X[-1].reshape(1, window, 1)
pred = model.predict(test_input)

print("Predicted next step count:", pred[0][0])
```

```python
# 2 Write a short code snippet to demonstrate the vanishing gradient problem in a basic RNN by training it on a sequence of 0s and 1s, then print the gradients after each epoch.<br><br><em><strong>Hint:</strong> Use TensorFlow or PyTorch and compare the gradients at early vs late time steps.</em>


import tensorflow as tf


X = np.random.randint(0, 2, (50, 5, 1)).astype(np.float32)
y = np.random.randint(0, 2, (50, 1)).astype(np.float32)

model = tf.keras.Sequential([
    tf.keras.layers.SimpleRNN(4, input_shape=(5,1)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

opt = tf.keras.optimizers.SGD(0.1)
loss_fn = tf.keras.losses.BinaryCrossentropy()

for epoch in range(2):
    with tf.GradientTape() as tape:
        pred = model(X, training=True)
        loss = loss_fn(y, pred)

    grads = tape.gradient(loss, model.trainable_variables)

    print("\nEpoch", epoch+1, "Grad mean:", 
          [float(tf.reduce_mean(tf.abs(g))) for g in grads])

    opt.apply_gradients(zip(grads, model.trainable_variables))
```

4 Write a short explanation (3-4 lines) of what 'unrolling an LSTM through time' means, using an example from a music playlist app that predicts the next song a user might play.


Unrolling an LSTM through time” means expanding the recurrent network across each time step so we can see how it processes a sequence step by step. For example, in a music playlist app, the LSTM might process songs like “rock → jazz → pop,” with each step updating its memory of the user’s taste. Unrolling shows how the model carries this memory forward while predicting the next song. It essentially turns the looped structure into a chain for easier understanding and training.
