```python
# 1 
def neuron_output(input_value, weight, bias):
    output = input_value * weight + bias
    return output

# Example usage
result1 = neuron_output(5, 0.8, 2)
print("Output 1:", result1)

result2 = neuron_output(10, 0.5, -1)
print("Output 2:", result2)
```

```python
# 2 
def neuron_output(input_value, weight, bias):
    return input_value * weight + bias

likes = [50, 120, 80, 200, 150]

weight = 0.8
bias = 2

for like_count in likes:
    output = neuron_output(like_count, weight, bias)
    print(f"Likes: {like_count}, Output: {output}")
```

```python
# 5
def neuron_output(inputs, weights, bias):
    weighted_sum = 0

    for i in range(len(inputs)):
        weighted_sum += inputs[i] * weights[i]

    
    return weighted_sum + bias

inputs = [2, 4, 1]
weights = [0.6, 0.3, 0.1]
bias = 1

output = neuron_output(inputs, weights, bias)

print("Neuron Output:", output)
```
