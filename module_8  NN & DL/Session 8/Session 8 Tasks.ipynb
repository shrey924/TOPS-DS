```python
# 4

import matplotlib.pyplot as plt

for lr in [0.01, 0.1, 0.5]:
    x = 0
    y = []
    for i in range(20):
        y.append(x)
        x = x - lr * (2 * (x - 3))
    plt.plot(y, label=f"lr={lr}")

plt.legend()
plt.show()
```

```python
# 5 
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 4, 100)
y = x**4 - 3*x**3 + 2

plt.plot(x, y)
plt.scatter(0, 2, label="Local Min")
plt.scatter(2.25, -6.54, label="Global Min")
plt.legend()
plt.show()
```
