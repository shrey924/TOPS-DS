```python
# 3
import math

y_true = [1, 0, 1]
y_pred = [0.9, 0.2, 0.7]

bce = -(math.log(0.9) + math.log(0.8) + math.log(0.7)) / 3

print("BCE Loss:", round(bce, 4))
```

```python
# 4 
y_true = [0, 0, 1]
y_pred = [0.2, 0.3, 0.5]

loss = -(y_true[0]*math.log(y_pred[0]) + y_true[1]*math.log(y_pred[1]) + y_true[2]*math.log(y_pred[2]))
print("Categorical Cross-Entropy Loss:", round(loss, 4))
```
