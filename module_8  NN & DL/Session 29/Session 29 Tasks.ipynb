```python
# 2 
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing import image

img_path = "image.avif"  

img = image.load_img(img_path, target_size=(128, 128))
img_array = image.img_to_array(img)
img_array = img_array / 255.0


X = np.expand_dims(img_array, axis=0)
y = np.array([1])
y = tf.keras.utils.to_categorical(y, num_classes=3)


model = Sequential([Conv2D(filters=32,kernel_size=(3, 3),activation='relu',input_shape=(128, 128, 3)),

    MaxPooling2D(pool_size=(2, 2)),

    Flatten(),

    Dense(64, activation='relu'),

    Dense(3, activation='softmax')
])
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()
history = model.fit(X,y,epochs=10,verbose=1)
model.save("fashion_cnn.h5")
print("Model saved successfully!")
prediction = model.predict(X)
class_names = ['bag', 'shoe', 't_shirt']
predicted_class = class_names[np.argmax(prediction)]

print("Prediction:", predicted_class)
print("Confidence:", np.max(prediction) * 100, "%")
```

```python
history = model.fit(
    X,      # single image
    y,      # label
    epochs=5,
    verbose=0
)

for i in range(5):
    print(f"Epoch {i+1}: Accuracy = {history.history['accuracy'][i]:.4f}")
```

```python
history = model.fit(
    X,
    y,
    validation_data=(X, y),
    epochs=5,
    verbose=0
)

for i in range(5):
    print(f"Epoch {i+1}: "
          f"Train Acc = {history.history['accuracy'][i]:.4f}, "
          f"Val Acc = {history.history['val_accuracy'][i]:.4f}")
```
