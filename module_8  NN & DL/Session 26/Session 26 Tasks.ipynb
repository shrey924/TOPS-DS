```python
# 2 Build a simple LSTM model in Keras to predict the next value in your prepared Flipkart price sequence and train it for 10 epochs.
prices = np.array([999, 1050, 1020, 1100, 1150, 1200, 1180, 1250, 1300, 1280])

def create_sequences(data, time_steps=3):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:i+time_steps])
        y.append(data[i+time_steps])
    return np.array(X), np.array(y)

X, y = create_sequences(prices, 3)

X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(3, 1)))
model.add(Dense(1))  
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
history = model.fit(X, y, epochs=10, verbose=1)
test_input = np.array([1250, 1300, 1280])
test_input = test_input.reshape((1, 3, 1))
predicted_price = model.predict(test_input)

print("Predicted next price:", predicted_price[0][0])
```

```python
# 3 Evaluate your trained LSTM model by predicting the next price in the sequence, and print both the predicted and actual values for comparison.<br><br><em><strong>Hint:</strong> Use model.predict() and compare with the last actual value in your test set.</em>

last_input = X[-1]  # shape (3, 1)
last_input = last_input.reshape((1, 3, 1))
predicted_price = model.predict(last_input)

print("Predicted Price:", predicted_price[0][0])
```

```python
actual_price = y[-1]

print("Actual Price:", actual_price)
```

```python
print("\n--- Model Evaluation ---")
print("Predicted Price:", predicted_price[0][0])
print("Actual Price   :", actual_price)
error = abs(predicted_price[0][0] - actual_price)
print("Absolute Error :", error)
```

```python
# 5 Use ChatGPT to generate a Python function that prepares any sequence data (such as daily Spotify song play counts) for LSTM input, and explain in your own words how the function works.

import numpy as np

def prepare_lstm_data(sequence, time_steps=3):
    """
    Converts a 1D sequence into LSTM-ready input format.

    Parameters:
    sequence (list or np.array): Time-series data (e.g., Spotify daily plays)
    time_steps (int): Number of past steps used to predict next value

    Returns:
    X: Input sequences (samples, time_steps, 1)
    y: Target values (next step prediction)
    """

    X, y = [], []

    for i in range(len(sequence) - time_steps):
        X.append(sequence[i:i + time_steps])   # past window
        y.append(sequence[i + time_steps])     # next value

    X = np.array(X)
    y = np.array(y)

    # Reshape for LSTM: (samples, timesteps, features)
    X = X.reshape((X.shape[0], X.shape[1], 1))

    return X, y
```
