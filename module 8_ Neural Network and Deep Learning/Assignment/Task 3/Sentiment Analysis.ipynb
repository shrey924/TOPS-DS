```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, GRU, Dense, Dropout
```

```python
df = pd.read_csv("imdb-movies-dataset.csv")
df.head()
```

```python
df.info()
```

```python
df["sentiment"] = (df["Rating"] >= 7).astype(int)
print(df["sentiment"].value_counts())
```

```python
X = df["Review"].astype(str)
y = df["sentiment"]
```

```python

tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(X)
X_seq = tokenizer.texts_to_sequences(X)
```

```python
X_pad = pad_sequences(X_seq, maxlen=200)

print(X_pad.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pad,y,test_size=0.2,random_state=42)
```

```python
# Rnn

rnn_model = Sequential([
    Embedding(5000, 128, input_length=200),
    SimpleRNN(64),
    Dropout(0.5),
    Dense(1, activation="sigmoid")
])
```

```python
rnn_model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
rnn_model.fit(X_train, y_train, epochs=5, batch_size=64)
rnn_pred = (rnn_model.predict(X_test) > 0.5).astype(int)
rnn_accuracy = accuracy_score(y_test, rnn_pred)
print("RNN Accuracy:", rnn_accuracy)
```

```python
# LSTM

lstm_model = Sequential([
    Embedding(5000, 128, input_length=200),
    LSTM(64),
    Dropout(0.5),
    Dense(1, activation="sigmoid")
])
```

```python
lstm_model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
lstm_model.fit(X_train, y_train, epochs=5, batch_size=64)
lstm_pred = (lstm_model.predict(X_test) > 0.5).astype(int)
lstm_accuracy = accuracy_score(y_test, lstm_pred)
print("LSTM Accuracy:", lstm_accuracy)
```

```python
# GRU 
gru_model = Sequential([
    Embedding(5000, 128, input_length=200),
    GRU(64),
    Dropout(0.5),
    Dense(1, activation="sigmoid")
])
```

```python
gru_model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
gru_model.fit(X_train, y_train, epochs=5, batch_size=64)
gru_pred = (gru_model.predict(X_test) > 0.5).astype(int)
gru_accuracy = accuracy_score(y_test, gru_pred)
print("GRU Accuracy:", gru_accuracy)
```

```python
results = pd.DataFrame({
    "Model": ["RNN", "LSTM", "GRU"],
    "Accuracy": [rnn_accuracy, lstm_accuracy, gru_accuracy]
})

print(results)
```

```python
best_model = results.loc[results["Accuracy"].idxmax()]

print("Best Model:", best_model["Model"])
print("Best Accuracy:", best_model["Accuracy"])
```
