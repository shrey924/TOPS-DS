```python
import cv2
import numpy as np
import os

img = cv2.imread(r"sheet.jpeg")
if img is None:
    print("Error: Image not found! Check the filename/path.")
else:
    print(f"Image loaded! Shape: {img.shape}")
```

```python
Image = cv2.resize(img, (520, 540))

print(f"Resized: {Image.shape}")
gray = cv2.cvtColor(Image, cv2.COLOR_BGR2GRAY)

print(f"Grayscale shape: {gray.shape}")
```

```python
noise = np.random.normal(0, 50, gray.shape)
noisy_img = gray + noise
noisy_image = np.clip(noisy_img, 0, 255)
noisy_image = noisy_image.astype(np.uint8)
cv2.imshow("noise", noisy_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
denoised = cv2.GaussianBlur(noisy_image, (3, 3), 0)

cv2.imshow("Gaussian Blur",denoised)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 5. Risks and ethical concerns

1. **Misinformation and fake images**
   Text-to-image models can create realistic fake photos that may mislead people on social media. Developers can reduce this risk by adding visible or invisible watermarks and using AI-content detection labels.

2. **Bias and harmful stereotypes**
   AI models may generate biased images based on gender, race, culture, or profession. Developers can reduce this risk by improving training data diversity, testing outputs for bias, and adding safety filters before publishing generated images.
