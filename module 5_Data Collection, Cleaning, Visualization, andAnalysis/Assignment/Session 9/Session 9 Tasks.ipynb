```python
# 1 
import pandas as pd

# List of delivery timestamps as strings
delivery_times = [
    '2024-06-01 14:30',
    '2024-06-02 09:15',
    '2024-06-03 20:45'
]

# Create DataFrame
df = pd.DataFrame({'delivery_timestamp': delivery_times})

print("Original DataFrame:")
print(df)

# Convert to datetime
df['delivery_timestamp'] = pd.to_datetime(df['delivery_timestamp'])

print("\nAfter Conversion:")
print(df)

print("\nData Type:")
print(df['delivery_timestamp'].dtype)
```

```python
# 2
import pandas as pd

# Create a dummy Flipkart orders dataset
data = {
    'order_id': [101, 102, 103, 104, 105, 106],
    'customer_name': ['Amit', 'Priya', 'Rahul', 'Sneha', 'Karan', 'Neha'],
    'order_date': [
        '2024-06-01 14:30:00',
        '2024-07-15 09:45:00',
        '2024-08-20 18:20:00',
        '2024-09-05 11:10:00',
        '2024-10-12 16:55:00',
        '2024-11-25 20:05:00'
    ],
    'amount': [1299, 2499, 999, 3499, 1799, 4599]
}

# Create DataFrame
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Convert order_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Extract year, month, and weekday
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['weekday'] = df['order_date'].dt.day_name()

print("\nUpdated DataFrame:")
print(df)
```

```python
# 3 
import pandas as pd

# Dummy Flipkart orders dataset
data = {
    'order_id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'order_date': [
        '2024-06-01 14:30:00',
        '2024-06-03 09:45:00',
        '2024-06-05 18:20:00',
        '2024-06-08 11:10:00',
        '2024-06-10 16:55:00',
        '2024-06-12 20:05:00',
        '2024-06-15 13:40:00',
        '2024-06-18 10:15:00',
        '2024-06-20 19:30:00',
        '2024-06-23 08:45:00'
    ]
}


df = pd.DataFrame(data)
df['order_date'] = pd.to_datetime(df['order_date'])
df = df.set_index('order_date')
weekly_orders = df.resample('W').size()

print("Orders Per Week:")
print(weekly_orders)
```

```python
# 4 
import pandas as pd

# Dummy Instagram posts dataset
data = {
    'post_id': [101, 102, 103, 104, 105, 106],
    'posted_at': [
        '2024-06-01 08:00:00',
        '2024-06-01 12:30:00',
        '2024-06-02 15:45:00',
        '2024-06-03 18:20:00',
        '2024-06-04 21:10:00',
        '2024-06-05 10:05:00'
    ]
}

df = pd.DataFrame(data)


df['posted_at'] = pd.to_datetime(df['posted_at'])
df['posted_at_kolkata'] = (
    df['posted_at']
    .dt.tz_localize('UTC')
    .dt.tz_convert('Asia/Kolkata')
)
print(df[['posted_at', 'posted_at_kolkata']].head())
```
