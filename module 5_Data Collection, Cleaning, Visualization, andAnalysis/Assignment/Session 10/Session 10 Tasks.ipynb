```python
# 2 
import pandas as pd
import matplotlib.pyplot as plt

# Dummy restaurant dataset
data = {
    'restaurant': [
        'Spice Hub', 'Urban Tadka', 'Pizza Point', 'Food Factory',
        'Royal Thali', 'Cafe Delight', 'Burger Zone',
        'Green Garden', 'Taste of India', 'Street Bites'
    ],
    'rating': [4.5, 4.2, 4.0, 3.8, 4.7, 4.1, 3.9, 4.4, 4.6, 3.7],
    'avg_meal_price': [800, 650, 500, 450, 900, 550, 400, 700, 850, 300]
}

df = pd.DataFrame(data)

# Scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df['rating'], df['avg_meal_price'])

plt.xlabel('Zomato Rating')
plt.ylabel('Average Meal Price (₹)')
plt.title('Restaurant Ratings vs Average Meal Price')

plt.grid(True)
plt.show()
```

```python
# 3 
import pandas as pd
import matplotlib.pyplot as plt

# Dummy data
data = {
    'platform': ['Swiggy', 'Zomato', "Domino's"],
    'orders': [18, 25, 12]
}

df = pd.DataFrame(data)

# Create bar chart
plt.figure(figsize=(6, 4))
plt.bar('Swiggy', 18, color='orange', label='Swiggy')
plt.bar('Zomato', 25, color='red', label='Zomato')
plt.bar("Domino's", 12, color='blue', label="Domino's")

# Labels and title
plt.xlabel('Food Delivery Platform')
plt.ylabel('Number of Orders')
plt.title('Orders Placed in the Last Month')

# Legend
plt.legend()

plt.show()
```

```python
# 4
import matplotlib.pyplot as plt

# Dummy data: Spotify listening session durations (in minutes)
session_durations = [
    15, 22, 30, 18, 45,
    50, 35, 40, 28, 60,
    55, 25, 20, 38, 42,
    48, 33, 27, 52, 65
]

# Create histogram
plt.figure(figsize=(8, 5))
plt.hist(session_durations, bins=5, color='skyblue', edgecolor='black')

# Labels and title
plt.xlabel('Session Duration (minutes)')
plt.ylabel('Frequency')
plt.title('Histogram of Spotify Listening Session Durations')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# 5
import matplotlib.pyplot as plt
import pandas as pd

# Dummy data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
screen_time = [45, 60, 55, 70, 80, 95, 75]  # Minutes per day
posts_liked = [12, 18, 15, 22, 30, 35, 25]

# Create figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Subplot 1: Line Plot for Screen Time
axes[0].plot(days, screen_time, marker='o', color='blue')
axes[0].set_title('Daily Instagram Screen Time')
axes[0].set_xlabel('Day')
axes[0].set_ylabel('Screen Time (minutes)')
axes[0].grid(True)

# Subplot 2: Bar Chart for Posts Liked
axes[1].bar(days, posts_liked, color='orange')
axes[1].set_title('Posts Liked Per Day')
axes[1].set_xlabel('Day')
axes[1].set_ylabel('Number of Likes')

plt.tight_layout()
plt.savefig('social_media_usage.png')
plt.show()

print("Figure saved as 'social_media_usage.png'")
```
