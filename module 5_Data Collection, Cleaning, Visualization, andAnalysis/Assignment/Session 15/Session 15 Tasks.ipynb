```python
# 2 
import pandas as pd
import sweetviz as sv

# Dummy Mumbai dataset
mumbai = pd.DataFrame({
    'restaurant_name': ['Cafe A', 'Cafe B', 'Cafe C', 'Cafe D', 'Cafe E'],
    'average_cost_for_two': [800, 1200, 900, 1500, 700],
    'user_rating': [4.3, 4.1, 4.5, 4.0, 4.2]
})

# Dummy Delhi dataset
delhi = pd.DataFrame({
    'restaurant_name': ['Rest X', 'Rest Y', 'Rest Z', 'Rest P', 'Rest Q'],
    'average_cost_for_two': [1000, 1400, 1100, 1800, 1300],
    'user_rating': [4.0, 3.8, 4.1, 3.9, 4.0]
})

# Save as CSV (optional)
mumbai.to_csv("zomato_mumbai.csv", index=False)
delhi.to_csv("zomato_delhi.csv", index=False)

# Create comparison report
report = sv.compare(
    [mumbai, "Mumbai Restaurants"],
    [delhi, "Delhi Restaurants"]
)

report.show_html("zomato_comparison_report.html")

print("Report generated: zomato_comparison_report.html")
```

```python
# 5
import pandas as pd

data = {
    "Order_ID": [1, 2, 3, 4, 5],
    "Restaurant": ["Dominos", "KFC", "Burger King", "Haldiram", "Subway"],
    "Item": ["Pizza", "Burger", "Whopper", "Thali", "Sandwich"],
    "Price": [299, 450, 399, 199, 249],
    "Rating": [4.2, 4.5, 4.0, 4.7, 4.1],
    "Delivery_Time(min)": [30, 40, 35, 25, 20]
}

df = pd.DataFrame(data)
df.to_csv("swiggy_orders.csv", index=False)

print("CSV file created successfully!")
```

```python
import pandas as pd
import sweetviz as sv

# Load dataset
df = pd.read_csv("swiggy_orders.csv")

# Create Sweetviz report
report = sv.analyze(df)

# Save report as HTML
report.show_html("swiggy_report.html")

print("Report generated: swiggy_report.html")
```
