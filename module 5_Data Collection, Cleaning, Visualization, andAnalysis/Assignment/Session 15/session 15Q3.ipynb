```python
import pandas as pd
import dtale

# -----------------------------
# Create Dummy Myntra Dataset
# -----------------------------
data = {
    "Product_ID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Brand": [
        "Nike", "Puma", "Adidas", "Roadster", "Levis",
        "HRX", "Allen Solly", "Tommy Hilfiger", "UCB", "Wrangler"
    ],
    "Product_Name": [
        "Running Shoes",
        "Sports T-Shirt",
        "Sneakers",
        "Casual Shirt",
        "Jeans",
        "Track Pants",
        "Formal Shirt",
        "Jacket",
        "Hoodie",
        "Denim Jacket"
    ],
    "Category": [
        "Footwear",
        "Clothing",
        "Footwear",
        "Clothing",
        "Clothing",
        "Clothing",
        "Clothing",
        "Clothing",
        "Clothing",
        "Clothing"
    ],
    "Price": [3500, 1200, 2800, 1800, 2500, 1500, 2200, 4500, 1999, 3000]
}

df = pd.DataFrame(data)

# Save CSV
df.to_csv("myntra_products.csv", index=False)

print("Dataset Created Successfully!")
print(df)

# -----------------------------
# Filter Products > 2000
# -----------------------------
filtered_df = df[df["Price"] > 2000]

print("\nProducts with Price > ₹2000")
print(filtered_df)

print("\nNumber of Products with Price > ₹2000:", len(filtered_df))

# -----------------------------
# Open D-Tale
# -----------------------------
d = dtale.show(df)
print("\nD-Tale URL:")
print(d._main_url)
```
