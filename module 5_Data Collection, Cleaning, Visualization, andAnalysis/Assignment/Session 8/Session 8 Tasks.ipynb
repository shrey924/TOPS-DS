```python
# 1 
import pandas as pd

# Dummy Zomato order history dataset
data = {
    'restaurant_name': [
        "McDonald's",
        "Domino's",
        "McDonald's",
        "Pizza Hut",
        "Subway",
        "Domino's",
        "KFC",
        "Burger King"
    ],
    'order_date': [
        '2026-06-20',
        '2026-06-21',
        '2026-06-20',  # Duplicate
        '2026-06-22',
        '2026-06-23',
        '2026-06-21',  # Duplicate
        '2026-06-24',
        '2026-06-25'
    ],
    'amount': [250, 450, 250, 550, 300, 450, 400, 350]
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)

# Find duplicate orders based on restaurant_name and order_date
duplicates = df[df.duplicated(
    subset=['restaurant_name', 'order_date'],
    keep=False
)]

print("\nDuplicate Orders:")
print(duplicates)
```

```python
# 2 
import pandas as pd

# Dummy Flipkart product reviews
reviews = [
    "Excellent product",
    "Good quality",
    "Excellent product",
    "Worth the money",
    "Good quality",
    "Excellent product",
    "Average performance",
    "Good quality",
    "Value for money",
    "Good quality",
    "Excellent product",
    "Battery backup is good"
]

# Create DataFrame
df = pd.DataFrame({'review_text': reviews})

print("Original Reviews:")
print(df)

# Count occurrences of each review
review_counts = df['review_text'].value_counts()

# Display top 3 most common duplicate reviews
top_3_duplicates = review_counts[review_counts > 1].head(3)

print("\nTop 3 Most Common Duplicate Reviews:")
print(top_3_duplicates)
```

```python
# 3 
import pandas as pd

# Mock Spotify playlists dataset
data = {
    'playlist_name': [
        'Chill Vibes',
        'Workout Hits',
        'Road Trip Songs',
        'Chill Vibes',      # Duplicate
        'Party Mix',
        'Workout Hits',     # Duplicate
        'Focus Beats',
        'Party Mix'         # Duplicate
    ],
    'creator_username': [
        'user123',
        'musiclover',
        'traveler99',
        'user123',          # Duplicate
        'djmax',
        'musiclover',       # Duplicate
        'studyguru',
        'djmax'             # Duplicate
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Remove exact duplicate rows
cleaned_df = df.drop_duplicates()

print("\nCleaned DataFrame (Duplicates Removed):")
print(cleaned_df)
```

```python
# 4 
import pandas as pd

# Dummy Instagram usernames dataset
df = pd.DataFrame({
    'username': [
        'insta_queen',
        'insta-queen',
        'instaqueen',
        'travel_guru',
        'insta_queen',
        'foodie123',
        'insta-queen'
    ]
})

print("Original DataFrame:")
print(df)

# Standardize username variants
df['username'] = df['username'].replace({
    'insta_queen': 'instaqueen',
    'insta-queen': 'instaqueen'
})

print("\nCleaned DataFrame:")
print(df)
```

```python
# 5 
import pandas as pd

# Dummy Paytm payment status data
df = pd.DataFrame({
    'user_id': [101, 102, 103, 104, 105, 106, 107, 108],
    'payment_status': ['Yes', ' yes ', 'Y', 'No', ' no ', 'N', 'YES', 'n']
})

print("Original DataFrame:")
print(df)

# Clean and standardize the payment_status column
df['payment_status'] = (
    df['payment_status']
    .str.strip()      # Remove leading/trailing spaces
    .str.lower()      # Convert to lowercase
    .replace({
        'yes': 1,
        'y': 1,
        'no': 0,
        'n': 0
    })
)

print("\nCleaned DataFrame:")
print(df)

print("\nData Types:")
print(df.dtypes)
```
