```python
# 1 
import matplotlib.pyplot as plt

# Sample Data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
sales = [100, 120, 90, 140, 160]

products = ['A', 'B', 'C', 'D']
quantities = [25, 40, 30, 20]

ratings = [3.5, 4.0, 4.2, 4.5, 4.8]
prices = [200, 350, 400, 550, 700]

categories = ['Food', 'Travel', 'Shopping', 'Bills']
expenses = [3000, 1500, 2500, 1000]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].plot(days, sales, marker='o')
axes[0, 0].set_title('Daily Sales')
axes[0, 0].set_xlabel('Day')
axes[0, 0].set_ylabel('Sales')

axes[0, 1].bar(products, quantities)
axes[0, 1].set_title('Product Quantities')
axes[0, 1].set_xlabel('Product')
axes[0, 1].set_ylabel('Quantity')

axes[1, 0].scatter(ratings, prices)
axes[1, 0].set_title('Rating vs Price')
axes[1, 0].set_xlabel('Rating')
axes[1, 0].set_ylabel('Price (₹)')

axes[1, 1].pie(expenses, labels=categories, autopct='%1.1f%%')
axes[1, 1].set_title('Monthly Expenses')

# Adjust layout
plt.tight_layout()

plt.show()
```

```python
# 3 
import matplotlib.pyplot as plt

# Dummy data
influencers = ['A', 'B', 'C', 'D', 'E']
followers = [120000, 95000, 150000, 80000, 175000]
posts = [2.5, 1.8, 3.2, 1.5, 4.0]

fig, ax1 = plt.subplots()

# Left y-axis: Followers
ax1.plot(influencers, followers, 'bo-', label='Followers')
ax1.set_ylabel('Followers', color='blue')

# Right y-axis: Daily Posts
ax2 = ax1.twinx()
ax2.plot(influencers, posts, 'rs--', label='Daily Posts')
ax2.set_ylabel('Avg Daily Posts', color='red')

plt.title('Followers vs Daily Posts')
plt.show()
```

```python
# 4 
import matplotlib.pyplot as plt

# Dummy data
movies = ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E']
tickets_sold = [50000, 75000, 60000, 90000, 70000]
imdb_ratings = [7.5, 8.2, 6.8, 9.0, 7.9]

# Scatter plot
plt.figure(figsize=(7, 5))
plt.scatter(tickets_sold, imdb_ratings)

# Add movie name annotations
for i in range(len(movies)):
    plt.annotate(
        movies[i],
        (tickets_sold[i], imdb_ratings[i]),
        xytext=(0, 5),
        textcoords='offset points'
    )

plt.xlabel('Tickets Sold')
plt.ylabel('IMDB Rating')
plt.title('Tickets Sold vs IMDB Rating')
plt.grid(True)

plt.show()
```

```python
# 5 
import matplotlib.pyplot as plt

# Dummy data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Oct']
sales = [120, 150, 180, 170, 200, 450]

fig, ax = plt.subplots()

# Line plot
ax.plot(months, sales, marker='o')

# Find highest sales point
max_sales = max(sales)
max_index = sales.index(max_sales)

# Add annotation
ax.annotate(
    'Big Billion Days',
    xy=(months[max_index], max_sales),
    xytext=(0, 15),
    textcoords='offset points',
    arrowprops=dict(arrowstyle='->')
)

ax.set_title('Flipkart Monthly Sales')
ax.set_xlabel('Month')
ax.set_ylabel('Sales')

plt.show()
```
