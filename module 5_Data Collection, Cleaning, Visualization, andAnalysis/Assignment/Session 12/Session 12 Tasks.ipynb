```python
# 1 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random delivery times for 50 orders
np.random.seed(42)
delivery_times = np.random.randint(15, 60, 50)

# Create DataFrame
df = pd.DataFrame({
    'delivery_time': delivery_times
})

# Create histplot
sns.histplot(data=df, x='delivery_time', bins=10, kde=True)

plt.title('Distribution of Zomato Delivery Times')
plt.xlabel('Delivery Time (minutes)')
plt.ylabel('Number of Orders')

plt.show()
```

```python
# 2
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Create boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(data=tips, x='day', y='total_bill')

# Customize chart
plt.title('Distribution of Total Bill Amount by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')

plt.show()
```

```python
# 3
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn theme
sns.set_theme(style="darkgrid")

# IPL Teams
teams = ['MI', 'CSK', 'RCB', 'KKR', 'DC', 'SRH', 'RR', 'PBKS']

# Simulate scores (20 matches per team)
np.random.seed(42)

data = []
for team in teams:
    scores = np.random.randint(120, 230, 20)
    for score in scores:
        data.append([team, score])

df = pd.DataFrame(data, columns=['Team', 'Runs'])

# Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Team', y='Runs')

plt.title('IPL Team Run Distributions')
plt.xlabel('Team')
plt.ylabel('Runs Scored')
plt.show()
```

```python
# 4



# Dummy Spotify track genres (40 songs)
genres = [
    'Pop', 'Rock', 'Hip-Hop', 'Jazz',
    'Pop', 'Rock', 'Pop', 'Jazz',
    'Hip-Hop', 'Pop', 'Rock', 'Jazz',
    'Pop', 'Hip-Hop', 'Rock', 'Pop',
    'Jazz', 'Pop', 'Rock', 'Hip-Hop',
    'Pop', 'Jazz', 'Rock', 'Pop',
    'Hip-Hop', 'Jazz', 'Pop', 'Rock',
    'Pop', 'Hip-Hop', 'Jazz', 'Rock',
    'Pop', 'Jazz', 'Hip-Hop', 'Rock',
    'Pop', 'Jazz', 'Rock', 'Hip-Hop'
]

# Create DataFrame
df = pd.DataFrame({'Genre': genres})

# Countplot
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Genre')

plt.title('Number of Songs per Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
# 5

# Daily step counts for a week (dummy fitness app data)
steps = [6500, 8200, 7500, 9100, 10500, 12000, 9800]

# Create DataFrame
df = pd.DataFrame({'Steps': steps})

# Apply whitegrid theme
sns.set_theme(style="whitegrid")

plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='Steps', fill=True, color='green')

plt.title('Distribution of Daily Step Counts')
plt.xlabel('Steps')
plt.ylabel('Density')

plt.show()
```
