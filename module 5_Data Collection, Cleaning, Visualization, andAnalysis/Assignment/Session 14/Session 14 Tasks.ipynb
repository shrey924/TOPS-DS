```python
# 1
import pandas as pd

# Dummy IPL match data
df = pd.DataFrame({
    'match_id': [1, 2, 3, 4, 5],
    'total_runs': [356, 298, 412, 325, 387]
})

print("Mean   :", df['total_runs'].mean())
print("Median :", df['total_runs'].median())
print("Min    :", df['total_runs'].min())
print("Max    :", df['total_runs'].max())
print("Std Dev:", df['total_runs'].std())
```

```python
# 2
import seaborn as sns
import matplotlib.pyplot as plt

# Dummy dataset
df = pd.DataFrame({
    'rating': [5,4,5,3,4,5,2,1,4,5,3,2,5,4,5,1,3,4,5,2],
    'category': [
        'Electronics','Fashion','Electronics','Books','Fashion',
        'Electronics','Books','Fashion','Electronics','Books',
        'Fashion','Electronics','Books','Fashion','Electronics',
        'Books','Fashion','Electronics','Books','Fashion'
    ]
})
sns.countplot(data=df, x='rating')
plt.title('Count of Flipkart Reviews by Rating')
plt.xlabel('Rating (Stars)')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
# 3


# Dummy Zomato data
df = pd.DataFrame({
    'average_cost_for_two': [500, 800, 1200, 1500, 700, 1000, 2000, 600],
    'user_rating': [3.8, 4.2, 4.5, 4.7, 3.9, 4.3, 4.8, 3.7]
})

sns.scatterplot(data=df,x='average_cost_for_two',y='user_rating')

plt.title('Cost vs User Rating')
plt.xlabel('Average Cost for Two (₹)')
plt.ylabel('User Rating')
plt.show()
```

```python
# 4


# Dummy dataset
df = pd.DataFrame({
    'genre': ['Action', 'Comedy', 'Drama', 'Action', 'Comedy',
              'Drama', 'Action', 'Thriller', 'Thriller', 'Drama'],
    'box_office_collection': [120, 80, 60, 150, 90,
                              70, 130, 110, 100, 65]
})

# Average box office collection by genre
avg_collection = (
    df.groupby('genre')['box_office_collection']
      .mean()
      .sort_values(ascending=False)
      .reset_index()
)

print(avg_collection)
```
