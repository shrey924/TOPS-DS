```python
# 4 
import pandas as pd
import numpy as np

# Simulated Zomato restaurant ratings dataset
df = pd.DataFrame({
    'Restaurant': ['Pizza Hub', 'Spice Garden', 'Burger Point',
                   'South Delight', 'Food Court', 'Cafe Treat'],
    'Rating': [4.2, np.nan, 4.5, np.nan, np.nan, 4.0]
})

print("Before Filling Missing Values:")
print(df)

# Forward fill missing ratings
df['Rating'] = df['Rating'].fillna(method='ffill')

# Backward fill any remaining missing ratings
df['Rating'] = df['Rating'].fillna(method='bfill')

print("\nAfter Forward Fill and Backward Fill:")
print(df)
```

The missing values in the Rating column are most likely MAR (Missing At Random). This is because the absence of ratings may be related to other factors in the dataset, such as a restaurant being newly added to Zomato or having very few customer reviews. In this case, the missingness depends on other observed information rather than being completely random.
