```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df['Marital_Status'].unique()
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({
    'Alone': 'Single',
    'YOLO': 'Single',
    'Absurd': 'Single'
})  
```

```python
df = pd.get_dummies(df, columns=['Marital_Status'])
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'],dayfirst = True)
```

```python
df['Dt_Customer'] = (pd.to_datetime('today') - df['Dt_Customer']).dt.days
```

```python
df['Dt_Customer'].unique()
```

```python
df.info()
```

```python
corr = df.corr()['Marital_Status_Widow']
```

```python
corr
```

```python
cols = corr[corr > 0.04].index
```

```python
new_df = df[cols]
```

```python
new_df = new_df.drop('Marital_Status_Widow',axis=1)
```

```python
new_df
```

```python
new_df.isnull().sum()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(new_df)
```

```python
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(new_df)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)
```

```python
new_df['Cluster'] = y_kmeans
```

```python
plt.scatter(X_scaled[:, 1], X_scaled[:, 2], c=y_kmeans)
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.title("Customer Segments")
plt.show()
```

```python
new_df.groupby('Cluster').mean()
```

2. Find Customer segments using Hierarchical clustering algorithm

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
dendrogram = sch.dendrogram(sch.linkage(X_scaled, method='ward'))

plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=5, linkage='ward')
y_hc = hc.fit_predict(X_scaled)

df['Cluster'] = y_hc
```

```python
df.groupby('Cluster').mean()
```
