```python

system_prompt = (
    "You are a Food Delivery Customer Support Agent. "
    "Your role is to assist customers professionally, politely, and helpfully. "
    "Provide accurate solutions based on the customer's issue. "
    "Keep every response under 80 words."
)

user_prompt_template = (
    "Customer Name: {customer_name}\n"
    "Order ID: {order_id}\n"
    "Issue Type: {issue_type}\n\n"
    "Please assist the customer with this issue."
)

ALLOWED_ISSUES = {"late delivery", "missing item", "wrong item"}

def validate_issue(issue_type):
    if issue_type not in ALLOWED_ISSUES:
        raise ValueError(
            f"Invalid issue_type '{issue_type}'. Allowed values are: late delivery, missing item, wrong item."
        )

def build_prompt(customer_name, order_id, issue_type):
    validate_issue(issue_type)
    user_prompt = user_prompt_template.format(
        customer_name=customer_name,
        order_id=order_id,
        issue_type=issue_type
    )
    print("=" * 60)
    print("SYSTEM PROMPT")
    print(system_prompt)
    print("\nUSER PROMPT")
    print(user_prompt)
    print("=" * 60)

test_cases = [
    {"customer_name": "Alice", "order_id": "FD12345", "issue_type": "late delivery"},
    {"customer_name": "Bob", "order_id": "FD67890", "issue_type": "missing item"},
    {"customer_name": "Charlie", "order_id": "FD54321", "issue_type": "cold food"},
]

for case in test_cases:
    try:
        build_prompt(**case)
    except ValueError as e:
        print(f"Error: Unable to create prompt. {e}")
```
