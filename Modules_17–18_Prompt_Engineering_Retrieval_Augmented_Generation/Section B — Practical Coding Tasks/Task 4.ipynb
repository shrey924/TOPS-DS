```python

import numpy as np
import faiss
from sentence_transformers import SentenceTransformer

policy_text = """Food delivery policies ensure customers receive fair service. Refunds are available when items are missing, incorrect, or damaged after verification. Customers should report issues within 24 hours of delivery. Full refunds are issued for missing items when confirmed, while partial refunds may apply for quality concerns. Delivery windows are estimates and may vary due to weather, traffic, or restaurant delays. Customers receive notifications if delays occur. Orders can be cancelled without charge before restaurant confirmation. After confirmation, cancellation fees may apply depending on preparation status. Customer support is available through chat and email for disputes. Food delivery policies ensure customers receive fair service. Refunds are available when items are missing, incorrect, or damaged after verification. Customers should report issues within 24 hours of delivery. Full refunds are issued for missing items when confirmed, while partial refunds may apply for quality concerns. Delivery windows are estimates and may vary due to weather, traffic, or restaurant delays. Customers receive notifications if delays occur. Orders can be cancelled without charge before restaurant confirmation. After confirmation, cancellation fees may apply depending on preparation status. Customer support is available through chat and email for disputes. Food delivery policies ensure customers receive fair service. Refunds are available when items are missing, incorrect, or damaged after verification. Customers should report issues within 24 hours of delivery. Full refunds are issued for missing items when confirmed, while partial refunds may apply for quality concerns. Delivery windows are estimates and may vary due to weather, traffic, or restaurant delays. Customers receive notifications if delays occur. Orders can be cancelled without charge before restaurant confirmation. After confirmation, cancellation fees may apply depending on preparation status. Customer support is available through chat and email for disputes. Food delivery policies ensure customers receive fair service. Refunds are available when items are missing, incorrect, or damaged after verification. Customers should report issues within 24 hours of delivery. Full refunds are issued for missing items when confirmed, while partial refunds may apply for quality concerns. Delivery windows are estimates and may vary due to weather, traffic, or restaurant delays. Customers receive notifications if delays occur. Orders can be cancelled without charge before restaurant confirmation. After confirmation, cancellation fees may apply depending on preparation status. Customer support is available through chat and email for disputes. Food delivery policies ensure customers receive fair service. Refunds are available when items are missing, incorrect, or damaged after verification. Customers should report issues within 24 hours of delivery. Full refunds are issued for missing items when confirmed, while partial refunds may apply for quality concerns. Delivery windows are estimates and may vary due to weather, traffic, or restaurant delays. Customers receive notifications if delays occur. Orders can be cancelled without charge before restaurant confirmation. After confirmation, cancellation fees may apply depending on preparation status. Customer support is available through chat and email for disputes. Food delivery policies ensure customers receive fair service. Refunds are available when items are missing, incorrect, or damaged after verification. Customers should report issues within 24 hours of delivery. Full refunds are issued for missing items when confirmed, while partial refunds may apply for quality concerns. Delivery windows are estimates and may vary due to weather, traffic, or restaurant delays. Customers receive notifications if delays occur. Orders can be cancelled without charge before restaurant confirmation. After confirmation, cancellation fees may apply depending on preparation status. Customer support is available through chat and email for disputes. Food delivery policies ensure customers receive fair service. Refunds are available when items are missing, incorrect, or damaged after verification. Customers should report issues within 24 hours of delivery. Full refunds are issued for missing items when confirmed, while partial refunds may apply for quality concerns. Delivery windows are estimates and may vary due to weather, traffic, or restaurant delays. Customers receive notifications if delays occur. Orders can be cancelled without charge before restaurant confirmation. After confirmation, cancellation fees may apply depending on preparation status. Customer support is available through chat and email for disputes. Food delivery policies ensure customers receive fair service. Refunds are available when items are missing, incorrect, or damaged after verification. Customers should report issues within 24 hours of delivery. Full refunds are issued for missing items when confirmed, while partial refunds may apply for quality concerns. Delivery windows are estimates and may vary due to weather, traffic, or restaurant delays. Customers receive notifications if delays occur. Orders can be cancelled without charge before restaurant confirmation. After confirmation, cancellation fees may apply depending on preparation status. Customer support is available through chat and email for disputes. """

def chunk_text(text, chunk_size=100, overlap=20):
    words=text.split()
    chunks=[]
    step=chunk_size-overlap
    for i in range(0,len(words),step):
        chunk=" ".join(words[i:i+chunk_size])
        if chunk:
            chunks.append(chunk)
    return chunks

chunks=chunk_text(policy_text)
print("Total chunks:",len(chunks))

model=SentenceTransformer("all-MiniLM-L6-v2")
emb=np.array(model.encode(chunks),dtype=np.float32)

index=faiss.IndexFlatL2(emb.shape[1])
index.add(emb)

def retrieve(query,k=3):
    q=np.array(model.encode([query]),dtype=np.float32)
    d,idx=index.search(q,k)
    return [chunks[i] for i in idx[0]]

def build_rag_prompt(query,retrieved_chunks):
    prompt="System: You are a food delivery policy assistant.\n\nContext:\n"
    for i,c in enumerate(retrieved_chunks,1):
        prompt+=f"{i}. {c}\n\n"
    prompt+=f"User Question: {query}\n\n"
    prompt+="Answer only from the provided context. If the answer is not present, say \'I don't know\'."
    return prompt

query="What is the refund policy for missing items?"
ctx=retrieve(query)

print("\nRetrieved Chunks:")
for i,c in enumerate(ctx,1):
    print(f"{i}. {c}\n")

prompt=build_rag_prompt(query,ctx)
print("\nAssembled Prompt:\n")
print(prompt)

print("\nSimulated Answer:")
print("Missing items are eligible for a full refund after verification if reported within 24 hours.")
```
