```python
examples = [
    ("My food arrived 45 minutes late.", "Late Delivery"),
    ("I ordered a pizza but received a burger.", "Wrong Item"),
    ("My drink was missing from the order.", "Missing Item"),
    ("The food was cold and tasted stale.", "Poor Quality"),
]

def build_few_shot_prompt(complaint_text):
   
    prompt = (
        "Classify the following food delivery complaint into one of these categories:\n"
        "Late Delivery, Wrong Item, Missing Item, Poor Quality.\n\n"
        "Examples:\n\n"
    )

    for text, label in examples:
        prompt += f"Input: {text}\n"
        prompt += f"Output: {label}\n\n"

    prompt += f"Input: {complaint_text}\n"
    prompt += "Output:"

    return prompt


def add_example(text, label):
    
    examples.append((text, label))

print("=" * 70)
print("PROMPT 1")
print(build_few_shot_prompt("My order reached almost an hour late."))
print()

print("=" * 70)
print("PROMPT 2")
print(build_few_shot_prompt("The fries were not included in my order."))
print()

add_example(
    "The dessert I ordered was missing.",
    "Missing Item"
)

print("=" * 70)
print("UPDATED PROMPT AFTER ADDING NEW EXAMPLE")
print(build_few_shot_prompt("The pasta was spoiled and smelled bad."))
```
