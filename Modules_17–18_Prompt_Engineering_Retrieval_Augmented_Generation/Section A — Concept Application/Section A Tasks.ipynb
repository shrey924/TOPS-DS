# Clear Prompt vs. Unclear Prompt

A **clear prompt** provides the LLM with specific instructions about its role, the decision rules it must follow, the information it should consider, and the format of the response. This helps the model produce accurate and consistent answers.

An **unclear prompt** is vague and does not define the conditions for making decisions or how the response should be structured. As a result, the LLM may interpret the request differently each time, leading to inconsistent refund approvals or rejections.

## Changes to Improve the Prompt

### 1. Define Explicit Decision Rules

The prompt should clearly state the conditions under which a refund can be approved or rejected.

**Example:**
- Approve a refund only if:
  - The order status is **Delivered**.
  - The refund request is made within **24 hours** of delivery.
  - The customer provides a valid reason.
- Otherwise, reject the request and explain why.

**Why this matters:**

Clear decision rules reduce ambiguity and ensure the chatbot follows the same logic for every customer instead of making random decisions.

---

### 2. Specify a Fixed Output Format

The prompt should require the chatbot to respond in a consistent structure.

**Example Output:**

- Decision: Approved / Rejected
- Reason:
- Next Steps:

**Why this matters:**

A fixed response format makes the chatbot's answers predictable, easier for users to understand, and simpler for developers to process or audit.

## Conclusion

By adding explicit decision rules and requiring a fixed output format, the chatbot becomes more reliable and consistent. These improvements reduce incorrect refund decisions and ensure every response follows the same reasoning process.

# Zero-Shot vs. Few-Shot Prompting for Complaint Classification

## Zero-Shot Prompting

Zero-shot prompting means giving the LLM only the task description without any examples. The model uses its pre-trained knowledge to classify customer complaints into the given categories.

**Advantages:**
- Uses very little context space.
- Fast and simple to implement.
- Suitable when the task is straightforward.

**Disadvantages:**
- May produce inconsistent classifications.
- Can misunderstand categories with similar meanings.
- Lower accuracy compared to few-shot prompting.

---

## Few-Shot Prompting

Few-shot prompting provides the LLM with a small number of labelled examples before asking it to classify a new complaint. The examples demonstrate the expected input-output pattern.

**Advantages:**
- Improves classification accuracy.
- Produces more consistent results.
- Helps the model understand the difference between similar categories.

**Disadvantages:**
- Uses more of the available context window.
- Too many examples may exceed the model's context limit.

---

## Which Approach Should Be Used?

Given the limited context window and 200 labelled examples, **few-shot prompting** is the better choice.

Instead of including all 200 examples, include only a small set of carefully selected examples that represent each category. This improves the model's understanding while staying within the context limit.

---

## Factors That Determine How Many Examples to Include

1. **Context Window Size**
   - The total prompt, including instructions, examples, and the customer message, must fit within the model's context limit.

2. **Task Complexity**
   - If the categories are similar or difficult to distinguish, more examples are needed.

3. **Category Coverage**
   - Include examples from all four categories: Late Delivery, Wrong Item, Missing Item, and Poor Quality.

4. **Example Quality**
   - A few clear and representative examples are more effective than many repetitive ones.

5. **Prompt Length**
   - The prompt should leave enough space for the new customer complaint and the model's response.

---

## Conclusion

Zero-shot prompting is efficient but may not provide consistent results. Few-shot prompting is the better approach because it improves accuracy and consistency while using only a small number of representative examples. The number of examples should be chosen based on the context window, task complexity, category coverage, and the quality of the examples.

# Chain-of-Thought (CoT) Prompting for Route Optimization

## How CoT Prompting Improves the Assistant's Output

Chain-of-Thought (CoT) prompting encourages the LLM to reason through a problem step by step before generating the final answer. Instead of directly suggesting the best delivery route, the model first analyzes the delivery priorities, traffic delays, and order sequence.

Using CoT prompting improves the assistant by:
- Increasing the accuracy of route planning through structured reasoning.
- Providing a clear explanation of how the optimal stop sequence was determined.
- Reducing inconsistent or random responses by following a logical process.

---

## Example of a CoT Prompt

```text
You are an AI assistant for delivery route planning.

Given five delivery addresses, their priority levels, and estimated traffic delays:

1. Analyze the priority level of each order.
2. Compare the traffic delays for each route.
3. Determine the most efficient delivery sequence while giving higher priority to urgent orders.
4. Explain your reasoning step by step.
5. Finally, provide the optimal stop sequence.

Input:
Address A – High Priority – 10 min delay
Address B – Medium Priority – 20 min delay
Address C – High Priority – 5 min delay
Address D – Low Priority – 15 min delay
Address E – Medium Priority – 8 min delay
```

---

## Limitation of CoT Prompting

One limitation of Chain-of-Thought prompting is that it generates longer responses, which increases token usage and response time. It may also not guarantee the correct answer if the model's reasoning process is flawed.

---

## Conclusion

Chain-of-Thought prompting improves route optimization by guiding the LLM through a logical reasoning process before producing the final route. This results in more accurate, transparent, and consistent recommendations, although it requires additional computation and longer responses.

# Why RAG is Better than Fine-Tuning for This Use Case

For an internal tool that allows delivery partners to query a **150-page policy handbook** that is updated every quarter, **Retrieval-Augmented Generation (RAG)** is a better choice than fine-tuning.

## Reason 1: Better Handling of Frequent Updates

The policy handbook is updated every quarter. With RAG, the latest version of the handbook only needs to be uploaded and re-indexed in the knowledge base. The LLM can immediately retrieve the updated information without retraining.

In contrast, fine-tuning requires retraining the model whenever the handbook changes, making the process slower and more expensive.

---

## Reason 2: More Accurate and Up-to-Date Responses

RAG retrieves the most relevant sections of the latest handbook before generating an answer. This ensures that responses are based on current policies.

A fine-tuned model stores knowledge in its parameters, so it may provide outdated information if it has not been retrained after policy updates.

---

## Handling Frequent Document Updates

| RAG | Fine-Tuning |
|------|-------------|
| Updates require only re-indexing the latest handbook. | Requires retraining the model after every handbook update. |
| Always retrieves the latest document content. | May return outdated information if not retrained. |
| Faster and lower maintenance cost. | Higher maintenance cost and longer update time. |

---

## When Fine-Tuning is a Better Choice

Fine-tuning is a better choice when the goal is to teach the model a **specific behavior, writing style, or specialized task** that does not change frequently.

**Example:** If the company wants the AI assistant to always respond in a consistent customer support tone or perform a specialized complaint classification task, fine-tuning would be more suitable because these behaviors remain stable over time.

---

## Conclusion

RAG is the preferred solution because it supports frequent document updates, provides accurate answers using the latest handbook, and is more cost-effective to maintain. Fine-tuning is better suited for learning permanent behaviors or domain-specific skills rather than storing frequently changing knowledge.

# Role of Chunk Size and Chunk Overlap in a RAG Pipeline

In a RAG (Retrieval-Augmented Generation) pipeline, documents are divided into smaller pieces called **chunks** before they are stored in a vector database. The quality of retrieval depends greatly on the **chunk size** and **chunk overlap**.

## Role of Chunk Size

Chunk size determines how much text is included in each chunk.

- **Small chunks** provide precise retrieval but may split related information across multiple chunks.
- **Large chunks** preserve more context but may include unnecessary information, reducing retrieval precision.

In this case, the dish name and its price are often separated into different chunks, causing incomplete retrieval.

---

## Role of Chunk Overlap

Chunk overlap means that consecutive chunks share some common text.

For example:

- Chunk 1: Lines 1–100
- Chunk 2: Lines 81–180

The overlapping lines help ensure that related information appearing near chunk boundaries is included in both chunks.

This reduces the chance of important details, such as a dish name and its price, being separated.

---

## How to Improve Retrieval Quality

### 1. Increase the Chunk Size

Use a larger chunk size so that related information, such as dish descriptions and prices, is more likely to remain in the same chunk.

### 2. Increase the Chunk Overlap

Increase the overlap between consecutive chunks (for example, 15–20% of the chunk size). This allows information near the boundaries to appear in multiple chunks, improving the chances of retrieving complete context.

---

## Trade-Off

Increasing the chunk size and chunk overlap improves retrieval quality but also increases storage requirements, indexing time, and retrieval cost. Larger or overlapping chunks may also introduce redundant information, leading to higher token usage during response generation.

---

## Conclusion

Chunk size controls how much information is stored in each chunk, while chunk overlap preserves context between adjacent chunks. For restaurant menu PDFs, increasing both the chunk size and overlap helps keep dish names and prices together, resulting in more accurate retrieval. However, this improvement comes at the cost of greater storage, processing time, and token usage.

# FAISS vs. ChromaDB as Vector Stores

Both **FAISS** and **ChromaDB** are popular vector stores used in RAG systems, but they have different strengths.

## FAISS

FAISS is a high-performance vector similarity search library designed for fast retrieval of embeddings.

**Advantages:**
- Very fast similarity search.
- Efficient for large datasets.
- Suitable for applications where search speed is the main priority.

**Best Scenario:**
FAISS is the better choice when the dataset is relatively stable and the system requires high-speed retrieval, such as searching millions of complaint embeddings.

---

## ChromaDB

ChromaDB is a vector database that stores embeddings along with metadata and provides built-in support for adding, updating, and deleting records.

**Advantages:**
- Easy management of embeddings and metadata.
- Supports document updates and deletion.
- Better integration with RAG frameworks like LangChain.

**Best Scenario:**
ChromaDB is ideal when documents are updated frequently and metadata filtering is required, such as maintaining a knowledge base that changes regularly.

---

## Would Switching to ChromaDB Solve the Outdated-Record Problem?

**Not completely.**

Although ChromaDB makes updating and deleting records easier than FAISS, simply switching vector stores will not automatically remove outdated complaint records. If old embeddings remain in the database, they can still be retrieved.

The root problem is that outdated records are not being properly updated or deleted from the vector store.

---

## Additional Step Required

Regardless of whether FAISS or ChromaDB is used, the system must implement a **document synchronization and index maintenance process**.

This includes:
- Removing embeddings for resolved or outdated complaints.
- Updating embeddings when complaint records change.
- Rebuilding or refreshing the index when necessary to ensure only current records are retrieved.

---

## Conclusion

FAISS is best for fast similarity search on relatively stable datasets, while ChromaDB is better for applications requiring frequent updates and metadata management. Switching to ChromaDB alone will not solve the outdated-record issue. A proper process for updating, deleting, and synchronizing embeddings is required regardless of the vector store used.
