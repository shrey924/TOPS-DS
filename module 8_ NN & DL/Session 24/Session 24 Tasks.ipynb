Why an RNN is better than a feedforward network?

A Recurrent Neural Network (RNN) is more suitable because step counts are sequential and time-dependent—today’s steps are influenced by previous days (like weekday routines, workouts, or fatigue patterns). An RNN can remember past information through its internal memory, while a standard feedforward neural network treats each input independently and ignores the order of the sequence.

```python
# 2  Implement a simple RNN cell from scratch in Python that takes a sequence of 5 numbers (e.g., song play counts per day from Spotify) and outputs the final hidden state after processing the sequence.<br><br><em><strong>Hint:</strong> Use the formula: h_t = tanh(Wx_t + Uh_{t-1} + b), where h_0 = 0.</em>

import numpy as np

# input sequence (e.g., Spotify play counts for 5 days)
x = np.array([3, 5, 2, 8, 6])

# initialize weights
Wx = 0.5
Uh = 0.8
b = 0.1

h = 0  # h0 = 0

# RNN computation
for xt in x:
    h = np.tanh(Wx * xt + Uh * h + b)

print("Final hidden state:", h)
```

```python
# 3 Given the following sentence: 'I love cricket and IPL matches', convert each word to a unique integer (word-to-index mapping), then simulate how an RNN would process this sequence by updating and printing the hidden state at each step.

import numpy as np

s = "I love cricket and IPL matches".split()
m = {w:i+1 for i,w in enumerate(s)}

Wx, Wh, b = 0.5, 0.8, 0.1
h = 0

for w in s:
    h = np.tanh(Wx*m[w] + Wh*h + b)
    print(w, round(h, 4))
```

4 Explain the vanishing gradient problem in RNNs using your own words, and give one real-world example (from text, speech, or time-series apps) where this issue might cause poor predictions.


The vanishing gradient problem in RNNs happens when the model tries to learn long-term dependencies, but the gradients (signals used to update weights during training) become extremely small as they are backpropagated through many time steps.

As a result, early parts of a sequence (like the beginning of a sentence) contribute very little to learning, because their influence “fades away” during training. This makes the RNN good at short-term patterns but poor at remembering long-range context.

```python
# 5  Use ChatGPT to generate a short Python code snippet that demonstrates Backpropagation Through Time (BPTT) for a 3-step RNN sequence. Paste the code and briefly describe what each part does.
import numpy as np
x = [1, 2, 3]
Wx, Wh, Wy = 0.5, 0.8, 1.0
h = 0
learning_rate = 0.01
hs = []
for t in x:
    h = np.tanh(Wx * t + Wh * h)
    hs.append(h)


y = Wy * hs[-1]

target = 1.0

loss = (y - target) ** 2

print("Loss:", loss)


dy = 2 * (y - target)

dWy = dy * hs[-1]

dh = dy * Wy

# backprop through time steps
for t in reversed(range(3)):
    dtanh = dh * (1 - hs[t] ** 2)

    dWx = dtanh * x[t]
    dWh = dtanh * (hs[t-1] if t > 0 else 0)

    print(f"Step {t+1} | dWx: {dWx:.4f}, dWh: {dWh:.4f}")
```
