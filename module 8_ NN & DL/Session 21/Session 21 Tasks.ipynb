```python
# 1 Create a new Keras Sequential model and add a Conv2D layer with 32 filters, a 3x3 kernel, and ReLU activation for image input shape (32, 32, 3) as used in CIFAR-10.

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D, Flatten, Dense, BatchNormalization, Dropout
from tensorflow.keras.utils import to_categorical
import numpy as np

model = Sequential()
model.add(Conv2D(filters=32,kernel_size=(3, 3),activation='relu',input_shape=(32, 32, 3)))
model.summary()
```

```python
# 2 Add a MaxPooling2D layer after your Conv2D layer to reduce spatial dimensions, then flatten the output and add a Dense layer with 64 units and ReLU activation.
from tensorflow.keras.models import Sequential

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(64, activation='relu')
])

model.summary()
```

```python
# 3 Insert a BatchNormalization layer after the Conv2D and before MaxPooling2D, and add a Dropout layer (rate 0.3) after flattening to help prevent overfitting.<br><br><em><strong>Hint:</strong> BatchNorm stabilizes training, Dropout randomly disables neurons during training.</em>



model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    BatchNormalization(),      # Normalize activations
    MaxPooling2D((2, 2)),      # Reduce spatial dimensions
    Flatten(),
    Dropout(0.3),              # Prevent overfitting
    Dense(64, activation='relu')
])

model.summary()
```

```python
# 4 Compile your model using the 'adam' optimizer, 'categorical_crossentropy' loss, and track 'accuracy' as a metric. Then, train the model for 5 epochs using random data shaped like CIFAR-10 images and labels.<br><br><em><strong>Constraint:</strong> Use numpy to generate random input data and one-hot encoded labels for testing if you don't want to download the full CIFAR-10 dataset.</em>

# Model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')  # 10 CIFAR-10 classes
])
```

```python
# Compile
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
X = np.random.rand(100, 32, 32, 3)
y = np.random.randint(0, 10, 100)
y = to_categorical(y, num_classes=10)
model.fit(X, y, epochs=5, batch_size=32)
print("Training Complete!")
```
