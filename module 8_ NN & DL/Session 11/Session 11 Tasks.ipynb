```python
# 1 
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.callbacks import LearningRateScheduler
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
model = Sequential([Flatten(input_shape=(28, 28)),Dense(128, activation='relu'),Dense(10, activation='softmax')])
```

```python
model.compile(optimizer='sgd',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train,epochs=5,validation_data=(X_test, y_test))
```

```python
print("Training Accuracy:", history.history['accuracy'][-1])
print("Training Loss:", history.history['loss'][-1])
print("Validation Accuracy:", history.history['val_accuracy'][-1])
print("Validation Loss:", history.history['val_loss'][-1])
```

```python
# 2
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
model = Sequential([Flatten(input_shape=(28, 28)),Dense(128, activation='relu'),Dense(10, activation='softmax')])
```

```python
model.compile(optimizer='rmsprop',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train,epochs=5,validation_data=(X_test, y_test))
```

```python
print(history.history['val_accuracy'][-1])
```

```python
# 3 
model = Sequential([Flatten(input_shape=(28, 28)),Dense(128, activation='relu'),Dense(10, activation='softmax')])
```

```python
# Adam with learning rate = 0.0005
model.compile(optimizer=Adam(learning_rate=0.0005),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train,epochs=5,validation_data=(X_test, y_test))
```

```python
print(history.history['val_accuracy'][-1])
```

```python
# 4
# Model
model = Sequential([Flatten(input_shape=(28,28)),Dense(128, activation='relu'),Dense(10, activation='softmax')])
```

```python
# Learning Rate Schedule
def schedule(epoch, lr):
    if epoch > 0 and epoch % 2 == 0:
        return lr * 0.5
    return lr

lr_scheduler = LearningRateScheduler(schedule)
```

```python
# SGD with Momentum
model.compile(optimizer=SGD(learning_rate=0.01, momentum=0.9),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train, y_train,epochs=6,validation_data=(X_test, y_test),callbacks=[lr_scheduler])
```

* Why Adam is Often Preferred

1 Faster Convergence
    Adam usually reaches good accuracy faster than SGD.
    It requires less manual tuning of the learning rate.

2 Adaptive Learning Rates
    Adam automatically adjusts the learning rate for each parameter.
    This helps the model learn efficiently even when features have different scales.

3 Works Well on Large Datasets
    Adam performs reliably on complex deep learning tasks such as image classification, recommendation systems, and NLP.

4 Combines Benefits of SGD and RMSProp
    It uses momentum (like SGD with Momentum) and adaptive learning rates (like RMSProp), making training more stable.
