```python
# 1 
import numpy as np

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
x = np.arange(-5, 6)

output = leaky_relu(x)

print("Input Values : ", x)
print("Output Values:", output)
```

```python
# 2


def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def parametric_relu(x, a):
    return np.where(x > 0, x, a * x)

x = np.linspace(-5, 5, 500)

y_relu = relu(x)
y_leaky = leaky_relu(x, alpha=0.01)
y_prelu = parametric_relu(x, a=0.2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))

plt.plot(x, y_relu, label='ReLU')
plt.plot(x, y_leaky, label='Leaky ReLU (α=0.01)')
plt.plot(x, y_prelu, label='PReLU (a=0.2)')

plt.title("ReLU vs Leaky ReLU vs Parametric ReLU")
plt.xlabel("Input (x)")
plt.ylabel("Output")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# 3 
import math

def softmax(scores):
    exp_values = [math.exp(x) for x in scores]
    total = sum(exp_values)

    probabilities = [x / total for x in exp_values]
    return probabilities

scores = [2.0, 1.0, 0.1, 0.5]
probs = softmax(scores)
print("Softmax Probabilities:")

for i, p in enumerate(probs):
    print(f"Class {i+1}: {p:.3f}")
```

```python
# 4 
def hard_sigmoid(x):
    return max(0, min(1, 0.2 * x + 0.5))
inputs = [-3, -1, 0, 1, 3]

for x in inputs:
    print(f"hard_sigmoid({x}) = {hard_sigmoid(x):.1f}")
```

(a) Hidden Dense Layer → ReLU
1 ReLU stands for Rectified Linear Unit.
2 It outputs 0 for negative values and the input itself for positive values.
3 Helps reduce the vanishing gradient problem.
4 Computationally simple and fast.

(b) Output Layer with 5 Classes → Softmax
1 Used for multi-class classification problems.
2 Converts raw output scores into probabilities.
3 All probabilities add up to 1.

(c) Output Layer with Binary Classification → Sigmoid
1 Used for binary classification problems.
2 Produces an output between 0 and 1.
3 The output can be interpreted as a probability.
4 Commonly used for decisions like:
5 Food vs Not Food
6 Trending vs Not Trending
7 Positive vs Negative
