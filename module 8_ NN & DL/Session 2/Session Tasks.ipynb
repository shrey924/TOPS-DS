1.Create a Python script using NumPy to implement a perceptron that learns to classify whether a Flipkart product review is 'positive' (1) or 'negative' (0) based on two features: review length and number of exclamation marks. Use a small manually created dataset for training.

```python
import numpy as np

X = np.array([[50, 5],   [45, 4],   [40, 3],   [10, 0],   [15, 1],   [20, 0]    ])
y = np.array([1, 1, 1, 0, 0, 0])
weights = np.zeros(X.shape[1])
bias = 0
learning_rate = 0.01
epochs = 100
```

```python
def step_function(x):
    return 1 if x >= 0 else 0
```

```python
# Training the perceptron
for epoch in range(epochs):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        prediction = step_function(linear_output)

        error = y[i] - prediction

        # Update weights and bias
        weights += learning_rate * error * X[i]
        bias += learning_rate * error
```

```python
print("Final Weights:", weights)
print("Final Bias:", bias)
```

```python
# 2 
def perceptron(inputs, weights, bias):
    weighted_sum = sum(i * w for i, w in zip(inputs, weights)) + bias
    if weighted_sum >= 0:
        return 1
    else:
        return 0
```

```python
# Case 1
inputs = [2, 3]
weights = [0.5, 0.4]
bias = -1

output = perceptron(inputs, weights, bias)
print("Case 1 Output:", output)
```

```python
# Case 2
inputs = [1, 1]
weights = [0.2, 0.3]
bias = -1

output = perceptron(inputs, weights, bias)
print("Case 2 Output:", output)
```

```python
# 4 
X = np.array([[0, 0],[0, 1],[1, 0],[1, 1]])
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
W_hidden = np.array([
    [20, -20],   # weights from x1
    [20, -20]    # weights from x2
])
b_hidden = np.array([-10, 30])
```

```python
W_output = np.array([[20],[20]])
b_output = np.array([-30])
```

```python
for x in X:

    # Hidden Layer
    hidden_input = np.dot(x, W_hidden) + b_hidden
    hidden_output = sigmoid(hidden_input)

    # Output Layer
    final_input = np.dot(hidden_output, W_output) + b_output
    final_output = sigmoid(final_input)

    prediction = 1 if final_output >= 0.5 else 0
```

```python
print("Input:", x)
print("Hidden Output:", np.round(hidden_output, 4))
print("Network Output:", np.round(final_output[0], 4))
print("Prediction:", prediction)
```
