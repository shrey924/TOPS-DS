```python
# 2 Create a Sequential model in Keras with three layers: an input Dense layer with 32 units and 'relu' activation, a Dense layer with 16 units and 'relu', and an output Dense layer with 1 unit and 'sigmoid'. Print the model summary.
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input, Concatenate
from tensorflow.keras.models import Model

model = Sequential([Dense(32, activation='relu', input_shape=(10,)),Dense(16, activation='relu'),Dense(1, activation='sigmoid')])

model.summary()
```

```python
# 3 Build a simple Keras Functional API model that takes a 10-feature input, passes it through one Dense layer with 20 units and 'relu', then outputs a single value with a Dense layer. Display the model summary and total parameter count.

inputs = Input(shape=(10,))
x = Dense(20, activation='relu')(inputs)
outputs = Dense(1)(x)

model = Model(inputs, outputs)
model.summary()

print("Total Parameters:", model.count_params())
```

```python
# 5 Use ChatGPT or Copilot to generate Keras code for a model with two inputs (e.g., one for user age and one for number of songs listened on Spotify) and one output (predicting likelihood of buying a premium subscription). Paste the generated code, run it, and note any errors or changes you had to make.

age_input = Input(shape=(1,), name="age")
songs_input = Input(shape=(1,), name="songs_listened")
```

```python
# Combine inputs
merged = Concatenate()([age_input, songs_input])
# Hidden layer
x = Dense(8, activation='relu')(merged)
# Output layer
output = Dense(1, activation='sigmoid')(x)
# Model
model = Model(inputs=[age_input, songs_input], outputs=output)
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model.summary()
```
