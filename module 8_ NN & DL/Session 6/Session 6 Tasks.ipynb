```python
# 1 
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
X = np.array([1.0, 2.0, 3.0])
W = np.array([
    [0.2, 0.4],
    [0.3, 0.5],
    [0.6, 0.1]
])
z = np.dot(X, W)
output = sigmoid(z)
print("Input Vector (X):")
print(X)
print("\nWeight Matrix (W):")
print(W)
print("\nDot Product (Z = X.W):")
print(z)
print("\nOutput After Sigmoid Activation:")
print(output)
```

```python
# 3 
import numpy as np
import matplotlib.pyplot as plt
def relu(x):
    return np.maximum(0, x)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
X = np.array([1.0, 2.0, 3.0])
W1 = np.array([
    [0.2, 0.4],
    [0.3, 0.5],
    [0.6, 0.1]
])
W2 = np.array([
    [0.7],
    [0.8]
])
Z1 = np.dot(X, W1)
A1 = relu(Z1)
Z2 = np.dot(A1, W2)
A2 = sigmoid(Z2)
```

```python
# Input Layer
plt.subplot(1, 3, 1)
plt.bar(range(len(X)), X)
plt.title("Input Layer")
plt.xlabel("Input Neurons")
plt.ylabel("Activation")
```

```python
# Hidden Layer
plt.subplot(1, 3, 2)
plt.bar(range(len(A1)), A1)
plt.title("Hidden Layer (ReLU)")
plt.xlabel("Hidden Neurons")
plt.ylabel("Activation")
```

```python
# Output Layer
plt.subplot(1, 3, 3)
plt.bar(range(len(A2)), A2)
plt.title("Output Layer (Sigmoid)")
plt.xlabel("Output Neurons")
plt.ylabel("Activation")
plt.tight_layout()
plt.show()
```

```python
# 4
relu = lambda x: np.maximum(0, x)
sigmoid = lambda x: 1 / (1 + np.exp(-x))
X = np.array([
    [1, 2],
    [2, 1],
    [3, 4]
])
np.random.seed(0)
W1 = np.random.randn(2, 3)  
W2 = np.random.randn(3, 1)  
hidden = relu(np.dot(X, W1))
output = sigmoid(np.dot(hidden, W2))
print("Output Activations:")
print(output)
```
