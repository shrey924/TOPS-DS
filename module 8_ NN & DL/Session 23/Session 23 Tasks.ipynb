```python
# 1 Load the ResNet50 model with pretrained ImageNet weights using Keras, print the model summary, and count how many layers are trainable by default.

from tensorflow.keras.applications import ResNet50


model = ResNet50(weights='imagenet') # Load pretrained ResNet50
model.summary() # Print model summary
trainable_layers = sum(layer.trainable for layer in model.layers) # Count trainable layers

print("\nTrainable Layers:", trainable_layers)
print("Total Layers:", len(model.layers))
```

```python
# 3 Freeze all layers of the InceptionV3 model except the last 30 layers, then add a custom Dense output layer for 5-class classification, and print the number of trainable parameters.<br><br><em><strong>Hint:</strong> Use model.trainable = False, then set trainable=True for the last layers as needed.</em>

from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Model

base = InceptionV3(weights='imagenet', include_top=False)

for layer in base.layers[:-30]:
    layer.trainable = False

x = GlobalAveragePooling2D()(base.output)
out = Dense(5, activation='softmax')(x)

model = Model(base.input, out)

print("Trainable Params:", sum(w.shape.num_elements() for w in model.trainable_weights))
```

```python
# 4 Use ChatGPT or Copilot to generate Keras code that loads the InceptionV3 model, freezes all layers, adds a custom classification head with a Dropout layer, and compiles the model for binary classification. Paste the generated code and note any changes you made to get it running.

from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model

# Load pretrained InceptionV3
base = InceptionV3(weights='imagenet', include_top=False, input_shape=(299, 299, 3))

# Freeze all layers
for layer in base.layers:
    layer.trainable = False

# Custom classification head
x = GlobalAveragePooling2D()(base.output)
x = Dropout(0.5)(x)
out = Dense(1, activation='sigmoid')(x)   # binary classification

model = Model(inputs=base.input, outputs=out)

# Compile model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```
