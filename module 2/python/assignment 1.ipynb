```python
# Write a Python program to find the first appearance of the substring 'not' and'poor' from a given string, if 'not' follows the 'poor', replace the whole 'not'...'poor' substring with 'good'. Return the resulting string

string = input("Enter a string: ")
not_pos = string.find('not')
poor_pos = string.find('poor')


if not_pos != -1 and poor_pos != -1 and not_pos < poor_pos:
    
    result = string[:not_pos] + 'good' + string[poor_pos + 4:]
else:
    result = string

print("Result:", result)



```

```python
# Program to find Greatest Common Divisor of two numbers. For example, theGCD of 20 and 28 is 4 and the GCD of 98 and 56 is 14.

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

print("The GCD of", num1, "and", num2, "is :", gcd(num1, num2))


```

```python
 # Counting the frequencies in a list using a dictionary in Python. Input : [1, 1, 1, 5, 5, 3, 1, 3, 3, 1,4, 4, 4, 2, 2, 2, 2]

nums = [1, 1, 1, 5, 5, 3, 1, 3, 3, 1, 4, 4, 4, 2, 2, 2, 2]

freq = {}

for n in nums:
    freq[n] = freq.get(n, 0) + 1  

for key, value in freq.items():
    print(f"{key} : {value}")


```

```python
# Write a python program using function to find the sum of odd series andevenseries 

def freq(n):
    sum_even = 0
    sum_odd = 0
    for i in range(n):
        if i % 2 == 0:
            sum_even += i
        else:
            sum_odd += i
    print(f"sum of even numbers: {sum_even}")
    print(f"sum of odd numbers: {sum_odd}")
num = int(input("Enter Any one number:"))
freq(num)



```

```python
# Python Program to Find Factorial of Number Using Recursion 

def factorial(n):
    if n == 0 or n == 1:   
        return 1
    else:
        return n * factorial(n - 1)   
num = int(input("Enter a number: "))
if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    print("The factorial of", num, "is:", factorial(num))





```

```python
# Write a Python function that takes a list and returns a new list with unique elements of the first list.

def unique_list(lst):
    unique = []
    for item in lst:
        if item not in unique:
            unique.append(item)
    return unique
numbers = [1, 2, 3, 2, 4, 1, 5]
print(unique_list(numbers))

```

```python
# Make a program to generate a strong password using the input given by theuser.To generate a password, randomly take some words from the user input andthen,
# include numbers, special characters and capital letters to generate the password. Also, keep a check that password length is more than 8 characters.
# Note: Include Exception handling wherever required. Also, make a ‘User’ classand store the details like user id, name and password of each user as a tuple




import random
import string

class User:
    def __init__(self, user_id, name, password):
        self.user_id = user_id
        self.name = name
        self.password = password

    def get_details(self):
        return (self.user_id, self.name, self.password)

def generate_password(text):
    words = text.split()
    if not words:
        raise ValueError("Please enter at least one word!")

    selected = random.choice(words)

    numbers = random.choice(string.digits)
    symbols = random.choice("!@#$%^&*")
    capital = random.choice(selected).upper()

    password = selected + capital + symbols + numbers 

    # Ensure password length > 8
    while len(password) < 8:
        password += random.choice(string.ascii_letters + string.digits)

    return password

try:
    user_id = input("Enter User ID: ")
    name = input("Enter Name: ")
    text = input("Enter some words: ")

    password = generate_password(text)
    user = User(user_id, name, password)

    print("\nUser Details (as Tuple):", user.get_details())
    print("Generated Password:", password)

except Exception as e:
    print("Error:", e)
```
